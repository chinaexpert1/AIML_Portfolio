1. Build a genome vector, score, and list the top Batman movies using similarity
metrics. Note that we will not use text processing on the movie title but rather similarity
vectors. Create a single similarity vector for Batman movies to be compared to other
genome vectors. One should use a distance measure between the Batman movie's
genome vector and the other movies' genome vectors. Use any approach as you see fit.

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize

# Load the data files
def load_movielens_data():
    """Load all necessary MovieLens data files"""
    # Load ratings
    ratings = pd.read_csv('movielens_rating.csv')
    
    # Load movies
    movies = pd.read_csv('movielens_movie.csv')
    
    # Load genome scores (movie-tag relevance scores)
    genome_scores = pd.read_csv('movielens_genome_scores.csv')
    
    # Load genome tags
    genome_tags = pd.read_csv('movielens_genome_tags.csv')
    
    return ratings, movies, genome_scores, genome_tags

# Create genome vectors for all movies
def create_genome_vectors(genome_scores, movies):
    """Create genome vectors for all movies"""
    # Pivot genome scores to create movie x tag matrix
    genome_matrix = genome_scores.pivot(index='movieId', 
                                       columns='tagId', 
                                       values='relevance')
    
    # Fill NaN values with 0
    genome_matrix = genome_matrix.fillna(0)
    
    # Merge with movie titles
    genome_df = genome_matrix.merge(movies[['movieId', 'title']], 
                                   left_index=True, 
                                   right_on='movieId', 
                                   how='inner')
    
    return genome_df

# Find Batman movies and create aggregate Batman genome vector
def create_batman_vector(genome_df):
    """Create a single genome vector representing all Batman movies"""
    # Find Batman movies
    batman_mask = genome_df['title'].str.contains('Batman', case=False, na=False)
    batman_movies = genome_df[batman_mask]
    
    print(f"Found {len(batman_movies)} Batman movies:")
    for idx, title in enumerate(batman_movies['title']):
        print(f"{idx+1}. {title}")
    
    # Get genome vectors for Batman movies (exclude movieId and title columns)
    batman_vectors = batman_movies.drop(['movieId', 'title'], axis=1).values
    
    # Create aggregate Batman vector (average of all Batman movie vectors)
    if len(batman_vectors) > 0:
        batman_aggregate_vector = np.mean(batman_vectors, axis=0)
    else:
        print("No Batman movies found!")
        return None, batman_movies
    
    return batman_aggregate_vector, batman_movies

# Calculate similarity and find top similar movies
def find_similar_movies(genome_df, batman_vector, top_n=20):
    """Find movies most similar to Batman movies using cosine similarity"""
    # Get all movie vectors
    movie_vectors = genome_df.drop(['movieId', 'title'], axis=1).values
    
    # Reshape batman_vector to 2D array for cosine_similarity
    batman_vector_2d = batman_vector.reshape(1, -1)
    
    # Calculate cosine similarity
    similarities = cosine_similarity(batman_vector_2d, movie_vectors)[0]
    
    # Add similarities to dataframe
    genome_df['batman_similarity'] = similarities
    
    # Sort by similarity and get top movies
    top_movies = genome_df.nlargest(top_n, 'batman_similarity')[['title', 'batman_similarity']]
    
    return top_movies

# Main execution
def main():
    # Load data
    print("Loading MovieLens data...")
    ratings, movies, genome_scores, genome_tags = load_movielens_data()
    
    # Create genome vectors
    print("\nCreating genome vectors...")
    genome_df = create_genome_vectors(genome_scores, movies)
    print(f"Created genome vectors for {len(genome_df)} movies")
    print(f"Genome vector dimensions: {len(genome_df.columns) - 2} tags")
    
    # Create Batman vector
    print("\n" + "="*50)
    print("Creating Batman genome vector...")
    batman_vector, batman_movies = create_batman_vector(genome_df)
    
    if batman_vector is None:
        return
    
    # Find similar movies
    print("\n" + "="*50)
    print("Finding movies similar to Batman movies using cosine similarity...")
    similar_movies = find_similar_movies(genome_df, batman_vector, top_n=50)
    
    # Rank Batman movies by similarity to aggregate Batman vector
    batman_ranked = similar_movies[similar_movies['title'].str.contains('Batman', case=False, na=False)]
    print("\nRanking of Batman movies by similarity to aggregate Batman genome vector:")
    print("="*70)
    for idx, (title, similarity) in enumerate(batman_ranked.values):
        print(f"{idx+1}. {title} (Similarity: {similarity:.4f})")
    
    print("\n" + "="*50)
    print("Top 20 movies similar to Batman (including Batman movies):")
    for idx, (title, similarity) in enumerate(similar_movies.head(20).values):
        print(f"{idx+1}. {title} (Similarity: {similarity:.4f})")
    
    # Filter out Batman movies to see other similar movies
    non_batman_similar = similar_movies[~similar_movies['title'].str.contains('Batman', case=False, na=False)]
    print("\n" + "="*50)
    print("Top 10 non-Batman movies similar to Batman:")
    for idx, (title, similarity) in enumerate(non_batman_similar.head(10).values):
        print(f"{idx+1}. {title} (Similarity: {similarity:.4f})")
    
    return genome_df, batman_vector, batman_movies, batman_ranked

# Run the analysis
if __name__ == "__main__":
    genome_df, batman_vector, batman_movies, batman_ranked = main()
```

2. List top-10 users who love (or scored high) Batman movies using the genome
vectors. Find the second most loved movie for these 10 users using the built pipeline.

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the data files
def load_movielens_data():
    """Load all necessary MovieLens data files"""
    # Load ratings
    ratings = pd.read_csv('movielens_rating.csv')
    
    # Load movies
    movies = pd.read_csv('movielens_movie.csv')
    
    # Load genome scores (movie-tag relevance scores)
    genome_scores = pd.read_csv('movielens_genome_scores.csv')
    
    # Load genome tags
    genome_tags = pd.read_csv('movielens_genome_tags.csv')
    
    return ratings, movies, genome_scores, genome_tags

# Create genome vectors (reusing from previous code)
def create_genome_vectors(genome_scores, movies):
    """Create genome vectors for all movies"""
    genome_matrix = genome_scores.pivot(index='movieId', 
                                       columns='tagId', 
                                       values='relevance')
    genome_matrix = genome_matrix.fillna(0)
    genome_df = genome_matrix.merge(movies[['movieId', 'title']], 
                                   left_index=True, 
                                   right_on='movieId', 
                                   how='inner')
    return genome_df

# Find users who love Batman movies
def find_batman_lovers(ratings, movies, genome_df, batman_vector):
    """Find top-10 users who rate Batman movies highly"""
    
    # Get Batman movie IDs
    batman_mask = movies['title'].str.contains('Batman', case=False, na=False)
    batman_movie_ids = movies[batman_mask]['movieId'].tolist()
    
    # Get all ratings for Batman movies
    batman_ratings = ratings[ratings['movieId'].isin(batman_movie_ids)]
    
    # Calculate each user's average rating for Batman movies
    user_batman_scores = batman_ratings.groupby('userId').agg({
        'rating': ['mean', 'count'],
        'movieId': 'nunique'
    }).reset_index()
    
    # Flatten column names
    user_batman_scores.columns = ['userId', 'avg_batman_rating', 'batman_ratings_count', 'batman_movies_watched']
    
    # Filter users who have rated at least 2 Batman movies (to ensure they're real Batman fans)
    user_batman_scores = user_batman_scores[user_batman_scores['batman_movies_watched'] >= 2]
    
    # Sort by average rating and get top 10
    top_batman_lovers = user_batman_scores.nlargest(10, 'avg_batman_rating')
    
    print("Top 10 Batman-loving users:")
    print("="*80)
    for idx, row in top_batman_lovers.iterrows():
        print(f"User {row['userId']}: Avg Batman Rating = {row['avg_batman_rating']:.2f}, "
              f"Rated {row['batman_movies_watched']} Batman movies")
    
    return top_batman_lovers

# Find second most loved movie using genome similarity
def find_second_most_loved_movies(top_users, ratings, movies, genome_df, batman_vector):
    """Find the second most loved movie for Batman-loving users using genome similarity"""
    
    results = []
    
    print("\n" + "="*80)
    print("Finding second most loved movies for each Batman-loving user...")
    print("="*80)
    
    for _, user_row in top_users.iterrows():
        user_id = user_row['userId']
        
        # Get all movies rated highly (>= 4.0) by this user
        user_high_ratings = ratings[(ratings['userId'] == user_id) & 
                                   (ratings['rating'] >= 4.0)]
        
        # Merge with movie info
        user_movies = user_high_ratings.merge(movies[['movieId', 'title']], on='movieId')
        
        # Get genome vectors for these movies
        user_movie_ids = user_movies['movieId'].tolist()
        user_genome_movies = genome_df[genome_df['movieId'].isin(user_movie_ids)]
        
        if len(user_genome_movies) == 0:
            continue
            
        # Calculate similarity to Batman vector for each movie
        movie_vectors = user_genome_movies.drop(['movieId', 'title'], axis=1).values
        batman_vector_2d = batman_vector.reshape(1, -1)
        similarities = cosine_similarity(batman_vector_2d, movie_vectors)[0]
        
        # Add similarity scores
        user_genome_movies = user_genome_movies.copy()
        user_genome_movies['batman_similarity'] = similarities
        
        # Merge with ratings
        user_genome_movies = user_genome_movies.merge(
            user_movies[['movieId', 'rating']], 
            on='movieId'
        )
        
        # Sort by similarity to Batman (descending)
        user_genome_movies = user_genome_movies.sort_values('batman_similarity', ascending=False)
        
        # Find non-Batman movies
        non_batman_movies = user_genome_movies[
            ~user_genome_movies['title'].str.contains('Batman', case=False, na=False)
        ]
        
        if len(non_batman_movies) > 0:
            # The "second most loved" movie is the non-Batman movie most similar to Batman
            second_loved = non_batman_movies.iloc[0]
            
            results.append({
                'userId': user_id,
                'avg_batman_rating': user_row['avg_batman_rating'],
                'second_loved_movie': second_loved['title'],
                'second_loved_rating': second_loved['rating'],
                'batman_similarity': second_loved['batman_similarity']
            })
            
            print(f"\nUser {user_id}:")
            print(f"  Second most loved movie: {second_loved['title']}")
            print(f"  Rating given: {second_loved['rating']}")
            print(f"  Batman similarity: {second_loved['batman_similarity']:.4f}")
    
    return pd.DataFrame(results)

# Main analysis
def main():
    # Load data
    print("Loading MovieLens data...")
    ratings, movies, genome_scores, genome_tags = load_movielens_data()
    
    # Create genome vectors
    print("\nCreating genome vectors...")
    genome_df = create_genome_vectors(genome_scores, movies)
    
    # Create Batman vector (aggregate)
    batman_mask = genome_df['title'].str.contains('Batman', case=False, na=False)
    batman_movies = genome_df[batman_mask]
    batman_vectors = batman_movies.drop(['movieId', 'title'], axis=1).values
    batman_vector = np.mean(batman_vectors, axis=0)
    
    print(f"\nFound {len(batman_movies)} Batman movies for analysis")
    
    # Find top Batman lovers
    print("\n" + "="*80)
    top_batman_lovers = find_batman_lovers(ratings, movies, genome_df, batman_vector)
    
    # Find their second most loved movies
    second_loved_results = find_second_most_loved_movies(
        top_batman_lovers, ratings, movies, genome_df, batman_vector
    )
    
    # Summary
    print("\n" + "="*80)
    print("SUMMARY: Second Most Loved Movies for Top 10 Batman Fans")
    print("="*80)
    print("\nThese are non-Batman movies that are most similar to Batman movies")
    print("and highly rated by users who love Batman:\n")
    
    for _, row in second_loved_results.iterrows():
        print(f"User {row['userId']}: {row['second_loved_movie']} "
              f"(Rating: {row['second_loved_rating']}, Similarity: {row['batman_similarity']:.3f})")
    
    # Analyze patterns
    print("\n" + "="*80)
    print("Most common 'second loved' movies among Batman fans:")
    movie_counts = second_loved_results['second_loved_movie'].value_counts()
    for movie, count in movie_counts.head().items():
        print(f"  {movie}: loved by {count} Batman fans")
    
    return top_batman_lovers, second_loved_results

# Run the analysis
if __name__ == "__main__":
    top_batman_lovers, second_loved_results = main()
```

3. Find the best genre for Batman movies. The genre is listed in the movie dataset
next to the movie title.

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
from collections import Counter


# Create genome vectors
def create_genome_vectors(genome_scores, movies):
    """Create genome vectors for all movies"""
    genome_matrix = genome_scores.pivot(index='movieId', 
                                       columns='tagId', 
                                       values='relevance')
    genome_matrix = genome_matrix.fillna(0)
    genome_df = genome_matrix.merge(movies[['movieId', 'title', 'genres']], 
                                   left_index=True, 
                                   right_on='movieId', 
                                   how='inner')
    return genome_df

# Parse genres (they're stored as pipe-separated values)
def parse_genres(genre_string):
    """Parse the pipe-separated genre string into a list"""
    if pd.isna(genre_string) or genre_string == '(no genres listed)':
        return []
    return genre_string.split('|')

# Analyze Batman movie genres
def analyze_batman_genres(movies):
    """Analyze the genres of Batman movies"""
    # Find Batman movies
    batman_mask = movies['title'].str.contains('Batman', case=False, na=False)
    batman_movies = movies[batman_mask].copy()
    
    print("Batman Movies and Their Genres:")
    print("="*80)
    
    # Parse genres for each Batman movie
    all_batman_genres = []
    for idx, row in batman_movies.iterrows():
        genres = parse_genres(row['genres'])
        all_batman_genres.extend(genres)
        print(f"{row['title']}: {', '.join(genres)}")
    
    # Count genre frequencies
    genre_counts = Counter(all_batman_genres)
    
    print("\n" + "="*80)
    print("Genre Distribution in Batman Movies:")
    print("="*80)
    total_genre_occurrences = sum(genre_counts.values())
    
    for genre, count in genre_counts.most_common():
        percentage = (count / len(batman_movies)) * 100
        print(f"{genre}: {count} occurrences ({percentage:.1f}% of Batman movies)")
    
    return batman_movies, genre_counts

# Find best genre using similarity approach
def find_best_genre_by_similarity(genome_df, batman_vector, top_n=100):
    """Find the best genre by analyzing genres of movies most similar to Batman"""
    
    # Calculate similarity to Batman vector for all movies
    movie_vectors = genome_df.drop(['movieId', 'title', 'genres'], axis=1).values
    batman_vector_2d = batman_vector.reshape(1, -1)
    similarities = cosine_similarity(batman_vector_2d, movie_vectors)[0]
    
    # Add similarities to dataframe
    genome_df = genome_df.copy()
    genome_df['batman_similarity'] = similarities
    
    # Get top N most similar movies
    top_similar = genome_df.nlargest(top_n, 'batman_similarity')
    
    # Analyze genres of these similar movies
    all_similar_genres = []
    genre_similarity_scores = {}
    
    for idx, row in top_similar.iterrows():
        genres = parse_genres(row['genres'])
        similarity = row['batman_similarity']
        
        # Weight each genre by the movie's similarity to Batman
        for genre in genres:
            if genre not in genre_similarity_scores:
                genre_similarity_scores[genre] = []
            genre_similarity_scores[genre].append(similarity)
        
        all_similar_genres.extend(genres)
    
    # Calculate average similarity score for each genre
    genre_avg_similarity = {}
    for genre, scores in genre_similarity_scores.items():
        genre_avg_similarity[genre] = np.mean(scores)
    
    # Count genre frequencies in similar movies
    similar_genre_counts = Counter(all_similar_genres)
    
    print("\n" + "="*80)
    print(f"Genre Distribution in Top {top_n} Movies Similar to Batman:")
    print("="*80)
    
    # Create a combined score (frequency * average similarity)
    genre_combined_scores = {}
    
    for genre in similar_genre_counts:
        frequency = similar_genre_counts[genre]
        avg_similarity = genre_avg_similarity[genre]
        combined_score = frequency * avg_similarity
        genre_combined_scores[genre] = {
            'frequency': frequency,
            'avg_similarity': avg_similarity,
            'combined_score': combined_score
        }
    
    # Sort by combined score
    sorted_genres = sorted(genre_combined_scores.items(), 
                          key=lambda x: x[1]['combined_score'], 
                          reverse=True)
    
    for genre, scores in sorted_genres[:10]:
        print(f"{genre}:")
        print(f"  Frequency: {scores['frequency']} movies")
        print(f"  Avg Similarity: {scores['avg_similarity']:.4f}")
        print(f"  Combined Score: {scores['combined_score']:.2f}")
    
    return sorted_genres, top_similar

# Compare with Batman movie genres
def compare_genre_analysis(batman_genre_counts, similar_genre_scores):
    """Compare genres from Batman movies vs similar movies"""
    
    print("\n" + "="*80)
    print("BEST GENRE ANALYSIS - Comparing Methods:")
    print("="*80)
    
    # Get top genres from Batman movies
    batman_top_genres = [genre for genre, _ in batman_genre_counts.most_common(5)]
    
    # Get top genres from similar movies
    similar_top_genres = [genre for genre, _ in similar_genre_scores[:5]]
    
    print("\nTop genres in actual Batman movies:")
    for i, genre in enumerate(batman_top_genres, 1):
        count = batman_genre_counts[genre]
        print(f"{i}. {genre} ({count} Batman movies)")
    
    print("\nTop genres in movies similar to Batman (weighted by similarity):")
    for i, (genre, scores) in enumerate(similar_genre_scores[:5], 1):
        print(f"{i}. {genre} (score: {scores['combined_score']:.2f})")
    
    # Find the best genre (appears high in both lists)
    all_top_genres = set(batman_top_genres) | set([g for g, _ in similar_genre_scores[:5]])
    
    print("\n" + "="*80)
    print("CONCLUSION - Best Genre for Batman Movies:")
    print("="*80)
    
    # The best genre should be prominent in both Batman movies and similar movies
    if 'Action' in batman_top_genres and 'Action' in similar_top_genres:
        best_genre = 'Action'
    elif similar_genre_scores[0][0] in batman_top_genres:
        best_genre = similar_genre_scores[0][0]
    else:
        best_genre = batman_top_genres[0]
    
    print(f"\nThe best genre for Batman movies is: {best_genre}")
    print("\nReasoning:")
    print(f"- Most common in Batman movies: {batman_top_genres[0]}")
    print(f"- Highest scoring in similar movies: {similar_genre_scores[0][0]}")
    print(f"- {best_genre} captures the essence of Batman movies best")
    
    return best_genre

# Main analysis
def main():
    # Load data
    print("Loading MovieLens data...")
    ratings, movies, genome_scores, genome_tags = load_movielens_data()
    
    # Create genome vectors
    print("\nCreating genome vectors...")
    genome_df = create_genome_vectors(genome_scores, movies)
    
    # Create Batman vector
    batman_mask = genome_df['title'].str.contains('Batman', case=False, na=False)
    batman_genome_movies = genome_df[batman_mask]
    batman_vectors = batman_genome_movies.drop(['movieId', 'title', 'genres'], axis=1).values
    batman_vector = np.mean(batman_vectors, axis=0)
    
    # Analyze Batman movie genres
    print("\n" + "="*80)
    batman_movies, batman_genre_counts = analyze_batman_genres(movies)
    
    # Find best genre using similarity
    similar_genre_scores, top_similar = find_best_genre_by_similarity(
        genome_df, batman_vector, top_n=100
    )
    
    # Compare and conclude
    best_genre = compare_genre_analysis(batman_genre_counts, similar_genre_scores)
    
    # Bonus: Show some non-Batman movies from the best genre that are similar to Batman
    print("\n" + "="*80)
    print(f"Examples of {best_genre} movies similar to Batman:")
    print("="*80)
    
    best_genre_movies = top_similar[
        (top_similar['genres'].str.contains(best_genre, na=False)) &
        (~top_similar['title'].str.contains('Batman', case=False, na=False))
    ].head(10)
    
    for idx, (_, row) in enumerate(best_genre_movies.iterrows(), 1):
        print(f"{idx}. {row['title']} (Similarity: {row['batman_similarity']:.4f})")
    
    return batman_movies, best_genre, similar_genre_scores

# Run the analysis
if __name__ == "__main__":
    batman_movies, best_genre, genre_analysis = main()
```

4. Extract features, X matrix: Utilize the user tags (df_tag Dataframe) to build item
vectors for clustering. The same approach can be used in the module with GloVe word
embeddings.

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix
import gensim.downloader
import re
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK data if needed
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
import numpy as np
import pandas as pd
import os

# Load GloVe embeddings from the extracted file
def load_glove_from_file(glove_file_path='glove_embeddings/glove.6B.100d.txt'):
    """
    Load GloVe embeddings from the extracted file
    """
    print(f"Loading GloVe embeddings from {glove_file_path}...")
    
    # Check if file exists
    if not os.path.exists(glove_file_path):
        print(f"ERROR: GloVe file not found at {glove_file_path}")
        print("Please make sure you:")
        print("1. Created a 'glove_embeddings' folder in your current directory")
        print("2. Extracted glove.6B.100d.txt into that folder")
        return None, None
    
    embeddings_dict = {}
    
    # Load the embeddings
    with open(glove_file_path, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            embeddings_dict[word] = vector
    
    print(f"Successfully loaded {len(embeddings_dict)} word vectors")
    print(f"Embedding dimension: {len(vector)}")
    
    # Create a class that mimics gensim's interface
    class GloveDict:
        def __init__(self, embeddings):
            self.embeddings = embeddings
            self.dim = 100
        
        def __contains__(self, word):
            return word in self.embeddings
        
        def __getitem__(self, word):
            if word in self.embeddings:
                return self.embeddings[word]
            else:
                # Return zero vector for unknown words
                return np.zeros(self.dim)
        
        def get(self, word, default=None):
            if word in self.embeddings:
                return self.embeddings[word]
            return default if default is not None else np.zeros(self.dim)
    
    return GloveDict(embeddings_dict), 100

# Load the embeddings
Glove, GLOVE_M = load_glove_from_file()

if Glove is not None:
    # Test the embeddings
    print("\nTesting GloVe embeddings:")
    test_words = ['movie', 'batman', 'action', 'superhero', 'dark', 'knight']
    for word in test_words:
        if word in Glove:
            vec = Glove[word]
            print(f"  '{word}': found ✓ (norm={np.linalg.norm(vec):.3f})")
        else:
            print(f"  '{word}': not in vocabulary")

    print("\nGloVe embeddings are ready to use!")
else:
    print("\nFailed to load GloVe embeddings. Check the file path.")



# Download punkt tokenizer if needed
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    print("Downloading NLTK punkt tokenizer...")
    nltk.download('punkt')

# The main function to create movie embeddings from user tags
def create_glove_embeddings(user_tags, movies, glove_model=Glove, embedding_dim=GLOVE_M):
    """Create movie embeddings using GloVe word embeddings"""
    
    if glove_model is None:
        print("ERROR: GloVe model not loaded!")
        return None, None
    
    print(f"\nCreating movie embeddings using GloVe (dim={embedding_dim})...")
    
    # Preprocess tags
    def preprocess_tag(tag):
        if not isinstance(tag, str):
            return []
        # Clean and tokenize
        tag = re.sub(r'[^\w\s]', ' ', tag.lower())
        tokens = word_tokenize(tag)
        return tokens
    
    # Group tags by movie
    movie_tag_groups = user_tags.groupby('movieId')['tag'].apply(list).reset_index()
    
    movie_embeddings = {}
    movies_with_valid_embeddings = []
    
    for _, row in movie_tag_groups.iterrows():
        movie_id = row['movieId']
        tags = row['tag']
        
        all_embeddings = []
        valid_words = []
        
        for tag in tags:
            tokens = preprocess_tag(tag)
            for token in tokens:
                if token in glove_model:
                    all_embeddings.append(glove_model[token])
                    valid_words.append(token)
        
        if all_embeddings:
            # Average all word embeddings for this movie
            movie_embedding = np.mean(all_embeddings, axis=0)
            movie_embeddings[movie_id] = movie_embedding
            movies_with_valid_embeddings.append({
                'movieId': movie_id,
                'valid_words': valid_words,
                'num_valid_words': len(valid_words)
            })
    
    # Convert to matrix
    movie_ids = list(movie_embeddings.keys())
    X_glove = np.array([movie_embeddings[mid] for mid in movie_ids])
    
    # Create dataframe with movie info
    glove_df = pd.DataFrame({'movieId': movie_ids})
    glove_df = glove_df.merge(movies[['movieId', 'title']], on='movieId', how='left')
    
    # Add statistics
    stats_df = pd.DataFrame(movies_with_valid_embeddings)
    glove_df = glove_df.merge(stats_df, on='movieId', how='left')
    
    print(f"Movie embedding matrix shape: {X_glove.shape}")
    print(f"Number of movies with valid embeddings: {len(movie_ids)}")
    print(f"Average words per movie: {glove_df['num_valid_words'].mean():.2f}")
    
    # Show examples
    print("\nExample movies with most GloVe words:")
    top_movies = glove_df.nlargest(5, 'num_valid_words')
    for _, row in top_movies.iterrows():
        print(f"  {row['title']}: {row['num_valid_words']} words")
        if row['valid_words']:
            sample_words = row['valid_words'][:5]
            print(f"    Sample words: {', '.join(sample_words)}")
    
    return X_glove, glove_df

# Example: Load data and create embeddings
def test_glove_embeddings():
    """Test function to verify everything works"""
    
    # Load movie data
    print("Loading movie data...")
    movies = pd.read_csv('movielens_movie.csv')
    user_tags = pd.read_csv('movielens_tag.csv')
    
    print(f"Loaded {len(movies)} movies and {len(user_tags)} user tags")
    
    # Create GloVe embeddings
    X_glove, glove_df = create_glove_embeddings(user_tags, movies)
    
    if X_glove is not None:
        print("\n✓ Success! GloVe embeddings created successfully.")
        print(f"We now have a matrix of shape {X_glove.shape} ready for clustering!")
        
        # Find similar movies example
        from sklearn.metrics.pairwise import cosine_similarity
        
        # Find a Batman movie
        batman_movies = glove_df[glove_df['title'].str.contains('Batman', case=False, na=False)]
        if len(batman_movies) > 0:
            batman_idx = batman_movies.index[0]
            batman_vec = X_glove[batman_idx].reshape(1, -1)
            
            # Calculate similarities
            similarities = cosine_similarity(batman_vec, X_glove).flatten()
            top_similar_idx = np.argsort(similarities)[-6:-1][::-1]  # Top 5 excluding itself
            
            print(f"\nMovies similar to '{batman_movies.iloc[0]['title']}':")
            for idx in top_similar_idx:
                if idx != batman_idx:
                    print(f"  - {glove_df.iloc[idx]['title']} (similarity: {similarities[idx]:.3f})")
    
    return X_glove, glove_df

# Run the test
if __name__ == "__main__" and Glove is not None:
    print("\n" + "="*60)
    print("Running test with our MovieLens data...")
    print("="*60)
    X_glove, glove_df = test_glove_embeddings()
```

5. Find out how close the two clustering approaches in the module and (Q4.) are to
the correct representation of the genres of the movies. The genres can also be represented
as the average word vectors based on genre terms. Note that the module uses only the first
term in the genre. A 2D visualization using PCA could help how these three representations
compare (title word embeddings, tag word embeddings, and genre word embeddings).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import confusion_matrix, adjusted_rand_score, normalized_mutual_info_score
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import entropy
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import cdist
import networkx as nx
from collections import Counter
import re

# ============================================
# EMBEDDING CREATION FUNCTIONS
# ============================================

def create_title_embeddings(movies, Glove, GLOVE_M=100):
    """Create embeddings from movie titles using GloVe"""
    print("Creating title embeddings...")
    
    # Create movie index
    movie_ids = sorted(movies['movieId'].unique())
    MoviesIndex = {mov: i for i, mov in enumerate(movie_ids)}
    M_titles = len(MoviesIndex)
    
    # Initialize arrays
    Titles = [None] * M_titles
    TitleGenres = [None] * M_titles
    TitlesParsed = [None] * M_titles
    
    # Populate titles and genres
    for _, row in movies.iterrows():
        if row['movieId'] in MoviesIndex:
            idx = MoviesIndex[row['movieId']]
            Titles[idx] = row['title']
            # Extract first genre only
            TitleGenres[idx] = re.search(r'[\w\-]+', row['genres']).group(0) if pd.notna(row['genres']) else 'Unknown'
            # Parse title into words
            TitlesParsed[idx] = re.findall(r'[a-z]+', row['title'].lower())
    
    # Create title embeddings
    X_titles = np.zeros((M_titles, GLOVE_M), dtype=np.float32)
    for idx, title_words in enumerate(TitlesParsed):
        if len(title_words) == 0:
            X_titles[idx] = np.zeros(GLOVE_M, dtype=np.float32)
        else:
            word_vectors = []
            for word in title_words:
                if word in Glove:
                    word_vectors.append(Glove[word])
                else:
                    word_vectors.append(np.zeros(GLOVE_M, dtype=np.float32))
            X_titles[idx] = np.mean(word_vectors, axis=0)
    
    # Create dataframe with movie info
    title_df = pd.DataFrame({
        'movieId': movie_ids,
        'title': Titles,
        'genre': TitleGenres
    })
    
    return X_titles, title_df, MoviesIndex

def create_tag_embeddings(user_tags, movies, Glove, GLOVE_M=100):
    """Create embeddings from user tags using GloVe"""
    print("Creating tag embeddings...")
    
    # Preprocess tags
    def preprocess_tag(tag):
        if not isinstance(tag, str):
            return []
        tag = re.sub(r'[^\w\s]', ' ', tag.lower())
        return tag.split()
    
    # Group tags by movie
    movie_tag_groups = user_tags.groupby('movieId')['tag'].apply(list).reset_index()
    
    movie_embeddings = {}
    
    for _, row in movie_tag_groups.iterrows():
        movie_id = row['movieId']
        tags = row['tag']
        
        all_embeddings = []
        
        for tag in tags:
            tokens = preprocess_tag(tag)
            for token in tokens:
                if token in Glove:
                    all_embeddings.append(Glove[token])
        
        if all_embeddings:
            movie_embeddings[movie_id] = np.mean(all_embeddings, axis=0)
    
    # Get movies that have both tags and are in the movie list
    movie_ids_with_tags = list(movie_embeddings.keys())
    
    # Create matrix
    X_tags = np.array([movie_embeddings[mid] for mid in movie_ids_with_tags])
    
    # Create dataframe
    tag_df = pd.DataFrame({'movieId': movie_ids_with_tags})
    tag_df = tag_df.merge(movies[['movieId', 'title', 'genres']], on='movieId', how='left')
    
    # Extract first genre
    tag_df['genre'] = tag_df['genres'].apply(
        lambda x: re.search(r'[\w\-]+', x).group(0) if pd.notna(x) else 'Unknown'
    )
    
    return X_tags, tag_df

def create_genre_embeddings(movies, Glove, GLOVE_M=100):
    """Create embeddings for each genre using GloVe"""
    print("Creating genre embeddings...")
    
    # Get all unique genres
    all_genres = []
    for genres_str in movies['genres'].dropna():
        genres = genres_str.split('|')
        all_genres.extend(genres)
    
    unique_genres = sorted(list(set(all_genres)))
    
    # Create embeddings for each genre
    genre_embeddings = {}
    
    for genre in unique_genres:
        # Parse genre name into words
        genre_words = re.findall(r'[a-z]+', genre.lower())
        
        if len(genre_words) == 0:
            genre_embeddings[genre] = np.zeros(GLOVE_M)
        else:
            word_vectors = []
            for word in genre_words:
                if word in Glove:
                    word_vectors.append(Glove[word])
                else:
                    word_vectors.append(np.zeros(GLOVE_M))
            
            genre_embeddings[genre] = np.mean(word_vectors, axis=0)
    
    return genre_embeddings, unique_genres

def create_genre_based_movie_embeddings(movies, genre_embeddings):
    """Create movie embeddings based on their genre embeddings"""
    print("Creating genre-based movie embeddings...")
    
    movie_genre_embeddings = []
    movie_ids = []
    movie_genres = []
    
    for _, row in movies.iterrows():
        if pd.notna(row['genres']):
            # Get all genres for this movie
            genres = row['genres'].split('|')
            
            # Get embeddings for each genre
            genre_vecs = []
            for genre in genres:
                if genre in genre_embeddings:
                    genre_vecs.append(genre_embeddings[genre])
            
            if genre_vecs:
                # Average genre embeddings for the movie
                movie_embedding = np.mean(genre_vecs, axis=0)
                movie_genre_embeddings.append(movie_embedding)
                movie_ids.append(row['movieId'])
                # Store first genre for evaluation
                movie_genres.append(genres[0])
    
    X_genre_based = np.array(movie_genre_embeddings)
    
    genre_based_df = pd.DataFrame({
        'movieId': movie_ids,
        'genre': movie_genres
    })
    
    return X_genre_based, genre_based_df

# ============================================
# VISUALIZATION FUNCTIONS
# ============================================

def three_way_comparison_visualization(X_titles, X_tags, X_genre, 
                                     title_df, tag_df, genre_df,
                                     n_sample=500):
    """Create comprehensive 3-way comparison visualization"""
    
    # Find common movies across all three methods
    common_movies = set(title_df['movieId']) & set(tag_df['movieId']) & set(genre_df['movieId'])
    print(f"Found {len(common_movies)} movies common to all three methods")
    
    # Get indices for common movies
    title_indices = [i for i, mid in enumerate(title_df['movieId']) if mid in common_movies]
    tag_indices = [i for i, mid in enumerate(tag_df['movieId']) if mid in common_movies]
    genre_indices = [i for i, mid in enumerate(genre_df['movieId']) if mid in common_movies]
    
    # Sample if too many
    if len(title_indices) > n_sample:
        sample_idx = np.random.choice(len(title_indices), n_sample, replace=False)
        title_indices = [title_indices[i] for i in sample_idx]
        tag_indices = [tag_indices[i] for i in sample_idx]
        genre_indices = [genre_indices[i] for i in sample_idx]
    
    # Get embeddings for common movies
    X_title_common = X_titles[title_indices]
    X_tag_common = X_tags[tag_indices]
    X_genre_common = X_genre[genre_indices]
    
    # Get genres for coloring
    genres_common = [title_df.iloc[i]['genre'] for i in title_indices]
    
    # Create figure with subplots
    fig = plt.figure(figsize=(20, 15))
    
    # 1. Individual PCA plots
    embeddings = {
        'Title Embeddings': X_title_common,
        'Tag Embeddings': X_tag_common,
        'Genre Embeddings': X_genre_common
    }
    
    # Get unique genres and assign colors
    unique_genres = sorted(list(set(genres_common)))
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_genres)))
    genre_to_color = {genre: colors[i] for i, genre in enumerate(unique_genres)}
    
    # Plot individual PCAs
    for idx, (name, X) in enumerate(embeddings.items()):
        ax = plt.subplot(2, 3, idx + 1)
        
        # Apply PCA
        pca = PCA(n_components=2, random_state=42)
        X_pca = pca.fit_transform(X)
        
        # Plot by genre
        for genre in unique_genres[:10]:  # Limit to 10 most common genres
            mask = [g == genre for g in genres_common]
            if sum(mask) > 0:
                X_genre = X_pca[mask]
                ax.scatter(X_genre[:, 0], X_genre[:, 1], 
                          c=[genre_to_color[genre]], 
                          label=genre, 
                          alpha=0.6, 
                          s=50)
        
        ax.set_title(f'{name}\n(Variance explained: {pca.explained_variance_ratio_.sum():.2%})', 
                    fontsize=12)
        ax.set_xlabel('PC1')
        ax.set_ylabel('PC2')
        
        if idx == 0:
            ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # 2. Combined PCA (all embeddings concatenated)
    ax = plt.subplot(2, 3, 4)
    
    # Concatenate all embeddings
    X_combined = np.hstack([X_title_common, X_tag_common, X_genre_common])
    
    # Apply PCA to combined
    pca_combined = PCA(n_components=2, random_state=42)
    X_combined_pca = pca_combined.fit_transform(X_combined)
    
    # Plot combined
    for genre in unique_genres[:10]:
        mask = [g == genre for g in genres_common]
        if sum(mask) > 0:
            X_genre = X_combined_pca[mask]
            ax.scatter(X_genre[:, 0], X_genre[:, 1], 
                      c=[genre_to_color[genre]], 
                      label=genre, 
                      alpha=0.6, 
                      s=50)
    
    ax.set_title(f'Combined Embeddings (All Methods)\n(Variance explained: {pca_combined.explained_variance_ratio_.sum():.2%})', 
                fontsize=12)
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    
    # 3. Similarity heatmap
    ax = plt.subplot(2, 3, 5)
    
    # Calculate average embeddings per genre for each method
    genre_avg_embeddings = {}
    
    for method_name, X in embeddings.items():
        genre_avg_embeddings[method_name] = {}
        for genre in unique_genres[:10]:
            mask = [g == genre for g in genres_common]
            if sum(mask) > 0:
                genre_avg_embeddings[method_name][genre] = np.mean(X[mask], axis=0)
    
    # Create correlation matrix between methods
    methods = list(embeddings.keys())
    correlation_matrix = np.zeros((len(methods), len(methods)))
    
    for i, method1 in enumerate(methods):
        for j, method2 in enumerate(methods):
            # Calculate average correlation across genres
            correlations = []
            for genre in unique_genres[:10]:
                if genre in genre_avg_embeddings[method1] and genre in genre_avg_embeddings[method2]:
                    corr = np.corrcoef(genre_avg_embeddings[method1][genre], 
                                      genre_avg_embeddings[method2][genre])[0, 1]
                    correlations.append(corr)
            
            correlation_matrix[i, j] = np.mean(correlations) if correlations else 0
    
    # Plot heatmap
    sns.heatmap(correlation_matrix, 
                xticklabels=methods, 
                yticklabels=methods, 
                annot=True, 
                fmt='.3f', 
                cmap='coolwarm', 
                center=0,
                ax=ax)
    ax.set_title('Average Correlation Between Methods\n(across genres)', fontsize=12)
    
    # 4. Distance analysis
    ax = plt.subplot(2, 3, 6)
    
    # Calculate how well each method separates genres
    separation_scores = {}
    
    for method_name, X in embeddings.items():
        # Calculate average intra-genre distance and inter-genre distance
        intra_distances = []
        inter_distances = []
        
        for genre in unique_genres[:5]:  # Use top 5 genres
            mask = [g == genre for g in genres_common]
            if sum(mask) > 1:
                X_genre = X[mask]
                # Intra-genre distances
                if len(X_genre) > 1:
                    intra_dist = np.mean(cdist(X_genre, X_genre, 'cosine'))
                    intra_distances.append(intra_dist)
                
                # Inter-genre distances
                other_mask = [g != genre for g in genres_common]
                if sum(other_mask) > 0:
                    X_other = X[other_mask]
                    inter_dist = np.mean(cdist(X_genre, X_other, 'cosine'))
                    inter_distances.append(inter_dist)
        
        avg_intra = np.mean(intra_distances) if intra_distances else 0
        avg_inter = np.mean(inter_distances) if inter_distances else 0
        separation_scores[method_name] = avg_inter - avg_intra
    
    # Plot separation scores
    methods = list(separation_scores.keys())
    scores = list(separation_scores.values())
    
    bars = ax.bar(methods, scores, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
    ax.set_title('Genre Separation Score\n(Higher = Better genre separation)', fontsize=12)
    ax.set_ylabel('Inter-genre distance - Intra-genre distance')
    ax.set_xticklabels(methods, rotation=45)
    
    # Add value labels on bars
    for bar, score in zip(bars, scores):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{score:.3f}',
                ha='center', va='bottom')
    
    plt.suptitle('Three-Way Comparison: Title, Tag, and Genre Embeddings', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    return separation_scores

# ============================================
# VISUALIZATION FUNCTIONS 
# ============================================

def create_confusion_matrix_visualization(clusters, true_genres, method_name):
    """Create an enhanced confusion matrix showing cluster-genre mapping"""
    
    # Get unique genres and clusters
    unique_genres = sorted(list(set(true_genres)))
    unique_clusters = sorted(list(set(clusters)))
    
    # Create confusion matrix
    conf_matrix = np.zeros((len(unique_clusters), len(unique_genres)))
    
    for cluster, genre in zip(clusters, true_genres):
        cluster_idx = unique_clusters.index(cluster)
        genre_idx = unique_genres.index(genre)
        conf_matrix[cluster_idx, genre_idx] += 1
    
    # Normalize by row (cluster) to show percentages
    conf_matrix_norm = conf_matrix / conf_matrix.sum(axis=1, keepdims=True)
    
    # Create figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    
    # Plot 1: Raw counts
    sns.heatmap(conf_matrix, 
                xticklabels=unique_genres, 
                yticklabels=[f'Cluster {i}' for i in unique_clusters],
                annot=True, 
                fmt='g', 
                cmap='Blues', 
                ax=ax1)
    ax1.set_title(f'{method_name} - Cluster vs Genre (Counts)')
    ax1.set_xlabel('True Genre')
    ax1.set_ylabel('Cluster')
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')
    
    # Plot 2: Normalized percentages
    sns.heatmap(conf_matrix_norm, 
                xticklabels=unique_genres, 
                yticklabels=[f'Cluster {i}' for i in unique_clusters],
                annot=True, 
                fmt='.2%', 
                cmap='RdYlBu_r', 
                vmin=0, 
                vmax=1,
                ax=ax2)
    ax2.set_title(f'{method_name} - Cluster vs Genre (Percentages)')
    ax2.set_xlabel('True Genre')
    ax2.set_ylabel('Cluster')
    plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right')
    
    plt.tight_layout()
    plt.show()
    
    return conf_matrix

def create_cluster_purity_visualization(clusters, true_genres, method_name):
    """Visualize cluster purity - how pure each cluster is in terms of genres"""
    
    # Calculate purity for each cluster
    cluster_purities = {}
    cluster_compositions = {}
    
    for cluster_id in set(clusters):
        # Get all genres in this cluster
        cluster_mask = clusters == cluster_id
        cluster_genres = [g for g, m in zip(true_genres, cluster_mask) if m]
        
        # Count genres
        genre_counts = Counter(cluster_genres)
        total = sum(genre_counts.values())
        
        # Purity is the fraction of the most common genre
        purity = max(genre_counts.values()) / total if total > 0 else 0
        cluster_purities[cluster_id] = purity
        
        # Store composition
        cluster_compositions[cluster_id] = {
            'genres': genre_counts,
            'total': total,
            'dominant_genre': max(genre_counts, key=genre_counts.get) if genre_counts else None
        }
    
    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot 1: Bar chart of cluster purities
    cluster_ids = sorted(cluster_purities.keys())
    purities = [cluster_purities[cid] for cid in cluster_ids]
    
    bars = ax1.bar(cluster_ids, purities, color=plt.cm.viridis(purities))
    ax1.set_xlabel('Cluster ID')
    ax1.set_ylabel('Purity Score')
    ax1.set_title(f'{method_name} - Cluster Purity Scores')
    ax1.set_ylim(0, 1.1)
    
    # Add value labels and dominant genre
    for i, (cid, purity) in enumerate(zip(cluster_ids, purities)):
        comp = cluster_compositions[cid]
        if comp['dominant_genre']:
            ax1.text(i, purity + 0.02, f"{purity:.2f}\n{comp['dominant_genre']}", 
                    ha='center', va='bottom', fontsize=8)
    
    # Plot 2: Stacked bar chart showing composition
    # Select top clusters by size
    top_clusters = sorted(cluster_compositions.items(), 
                         key=lambda x: x[1]['total'], 
                         reverse=True)[:10]
    
    # Get all genres across top clusters
    all_genres = set()
    for _, comp in top_clusters:
        all_genres.update(comp['genres'].keys())
    all_genres = sorted(list(all_genres))
    
    # Create stacked data
    bottom = np.zeros(len(top_clusters))
    colors = plt.cm.tab20(np.linspace(0, 1, len(all_genres)))
    
    for i, genre in enumerate(all_genres):
        values = [comp['genres'].get(genre, 0) / comp['total'] 
                 for _, comp in top_clusters]
        ax2.bar(range(len(top_clusters)), values, bottom=bottom, 
               label=genre, color=colors[i])
        bottom += values
    
    ax2.set_xlabel('Cluster (by size)')
    ax2.set_ylabel('Genre Proportion')
    ax2.set_title(f'{method_name} - Top 10 Cluster Compositions')
    ax2.set_xticks(range(len(top_clusters)))
    ax2.set_xticklabels([f'C{cid}\n({comp["total"]} movies)' 
                         for cid, comp in top_clusters])
    ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    plt.tight_layout()
    plt.show()
    
    # Calculate overall purity
    overall_purity = np.mean(list(cluster_purities.values()))
    print(f"{method_name} - Average Cluster Purity: {overall_purity:.3f}")
    
    return cluster_purities, cluster_compositions

def create_tsne_visualization(X_list, labels_list, method_names, true_genres):
    """Create t-SNE visualization comparing different embedding methods"""
    
    fig, axes = plt.subplots(1, len(X_list), figsize=(6*len(X_list), 5))
    if len(X_list) == 1:
        axes = [axes]
    
    # Get color mapping for genres
    unique_genres = sorted(list(set(true_genres)))[:15]  # Limit to 15 most common
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_genres)))
    genre_to_color = {genre: colors[i] for i, genre in enumerate(unique_genres)}
    
    for idx, (X, clusters, method_name) in enumerate(zip(X_list, labels_list, method_names)):
        # Apply t-SNE
        tsne = TSNE(n_components=2, random_state=42, perplexity=30)
        X_tsne = tsne.fit_transform(X[:1000])  # Limit to 1000 for speed
        genres_subset = true_genres[:1000]
        
        ax = axes[idx]
        
        # Plot by genre
        for genre in unique_genres:
            mask = [g == genre for g in genres_subset]
            if sum(mask) > 0:
                X_genre = X_tsne[mask]
                ax.scatter(X_genre[:, 0], X_genre[:, 1], 
                          c=[genre_to_color[genre]], 
                          label=genre, 
                          alpha=0.6, 
                          s=30)
        
        ax.set_title(f'{method_name} - t-SNE Visualization')
        ax.set_xlabel('t-SNE 1')
        ax.set_ylabel('t-SNE 2')
        
        if idx == 0:
            ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    plt.tight_layout()
    plt.show()

def create_clustering_quality_metrics(X_list, clusters_list, true_genres, method_names):
    """Compare multiple clustering quality metrics across methods"""
    
    # Convert genres to numeric labels
    genre_to_int = {g: i for i, g in enumerate(sorted(set(true_genres)))}
    true_labels = [genre_to_int[g] for g in true_genres]
    
    metrics_data = []
    
    for X, clusters, method_name in zip(X_list, clusters_list, method_names):
        # Calculate various metrics
        ari = adjusted_rand_score(true_labels, clusters)
        nmi = normalized_mutual_info_score(true_labels, clusters)
        silhouette = silhouette_score(X, clusters)
        calinski = calinski_harabasz_score(X, clusters)
        davies_bouldin = davies_bouldin_score(X, clusters)
        
        # Calculate homogeneity (how pure clusters are)
        homogeneity = 0
        for cluster_id in set(clusters):
            cluster_mask = clusters == cluster_id
            cluster_genres = [g for g, m in zip(true_genres, cluster_mask) if m]
            if cluster_genres:
                genre_counts = Counter(cluster_genres)
                total = sum(genre_counts.values())
                probs = [count/total for count in genre_counts.values()]
                homogeneity += entropy(probs) * total
        homogeneity = 1 - (homogeneity / len(clusters))
        
        metrics_data.append({
            'Method': method_name,
            'ARI': ari,
            'NMI': nmi,
            'Silhouette': silhouette,
            'Calinski-Harabasz': calinski / 1000,  # Scale for visualization
            'Davies-Bouldin': 1 / (1 + davies_bouldin),  # Invert so higher is better
            'Homogeneity': homogeneity
        })
    
    # Create radar chart
    fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(projection='polar'))
    
    # Set up angles
    metrics = ['ARI', 'NMI', 'Silhouette', 'Calinski-Harabasz', 'Davies-Bouldin', 'Homogeneity']
    angles = np.linspace(0, 2 * np.pi, len(metrics), endpoint=False).tolist()
    angles += angles[:1]  # Complete the circle
    
    # Plot each method
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
    for i, data in enumerate(metrics_data):
        values = [data[m] for m in metrics]
        values += values[:1]  # Complete the circle
        
        ax.plot(angles, values, 'o-', linewidth=2, label=data['Method'], color=colors[i])
        ax.fill(angles, values, alpha=0.25, color=colors[i])
    
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metrics)
    ax.set_ylim(0, 1)
    ax.set_title('Clustering Quality Metrics Comparison\n(Higher values are better for all metrics)', 
                 y=1.08, fontsize=14)
    ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
    ax.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Create bar chart for easier comparison
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Prepare data for grouped bar chart
    methods = [d['Method'] for d in metrics_data]
    x = np.arange(len(metrics))
    width = 0.25
    
    for i, method_data in enumerate(metrics_data):
        values = [method_data[m] for m in metrics]
        offset = (i - len(metrics_data)/2) * width + width/2
        bars = ax.bar(x + offset, values, width, label=method_data['Method'])
        
        # Add value labels
        for bar, val in zip(bars, values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height,
                    f'{val:.3f}', ha='center', va='bottom', fontsize=8)
    
    ax.set_xlabel('Metrics')
    ax.set_ylabel('Score')
    ax.set_title('Clustering Quality Metrics - Bar Chart Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics, rotation=45)
    ax.legend()
    ax.set_ylim(0, max([max([d[m] for m in metrics]) for d in metrics_data]) * 1.2)
    
    plt.tight_layout()
    plt.show()
    
    return pd.DataFrame(metrics_data)

def comprehensive_clustering_analysis(X_titles, X_tags, X_genre, 
                                    clusters_titles, clusters_tags, clusters_genre,
                                    title_df, tag_df, genre_df):
    """Run all visualizations for comprehensive analysis"""
    
    # Find common movies
    common_movies = set(title_df['movieId']) & set(tag_df['movieId']) & set(genre_df['movieId'])
    
    # Get indices for common movies
    title_indices = [i for i, mid in enumerate(title_df['movieId']) if mid in common_movies]
    tag_indices = [i for i, mid in enumerate(tag_df['movieId']) if mid in common_movies]
    genre_indices = [i for i, mid in enumerate(genre_df['movieId']) if mid in common_movies]
    
    # Get common data
    X_list = [X_titles[title_indices], X_tags[tag_indices], X_genre[genre_indices]]
    clusters_list = [clusters_titles[title_indices], clusters_tags[tag_indices], clusters_genre[genre_indices]]
    true_genres = [title_df.iloc[i]['genre'] for i in title_indices]
    method_names = ['Title-based', 'Tag-based', 'Genre-based']
    
    print("="*80)
    print("COMPREHENSIVE CLUSTERING ANALYSIS")
    print("="*80)
    
    # 1. Confusion matrices
    print("\n1. Confusion Matrix Analysis")
    for clusters, method_name in zip(clusters_list, method_names):
        create_confusion_matrix_visualization(clusters, true_genres, method_name)
    
    # 2. Cluster purity
    print("\n2. Cluster Purity Analysis")
    for clusters, method_name in zip(clusters_list, method_names):
        create_cluster_purity_visualization(clusters, true_genres, method_name)
    
    # 3. t-SNE visualization
    print("\n3. t-SNE Visualization")
    create_tsne_visualization(X_list, clusters_list, method_names, true_genres)
    
    # 4. Clustering quality metrics
    print("\n4. Clustering Quality Metrics")
    metrics_df = create_clustering_quality_metrics(X_list, clusters_list, true_genres, method_names)
    print("\nMetrics Summary:")
    print(metrics_df.to_string(index=False))
    
    return metrics_df

# ============================================
# MAIN EXECUTION FUNCTION
# ============================================

def run_complete_analysis(movies, user_tags, Glove, GLOVE_M=100):
    """Run the complete clustering analysis for Question 5"""
    
    print("="*80)
    print("QUESTION 5: CLUSTERING COMPARISON ANALYSIS")
    print("="*80)
    
    # Step 1: Create all three types of embeddings
    print("\nStep 1: Creating embeddings...")
    print("-"*40)
    
    # 1.1 Title embeddings
    X_titles, title_df, MoviesIndex = create_title_embeddings(movies, Glove, GLOVE_M)
    print(f"✓ Title embeddings created: shape {X_titles.shape}")
    
    # 1.2 Tag embeddings
    X_tags, tag_df = create_tag_embeddings(user_tags, movies, Glove, GLOVE_M)
    print(f"✓ Tag embeddings created: shape {X_tags.shape}")
    
    # 1.3 Genre embeddings
    genre_embeddings, unique_genres = create_genre_embeddings(movies, Glove, GLOVE_M)
    X_genre, genre_df = create_genre_based_movie_embeddings(movies, genre_embeddings)
    print(f"✓ Genre embeddings created: shape {X_genre.shape}")
    print(f"✓ Found {len(unique_genres)} unique genres")
    
    # Step 2: Perform clustering on each embedding type
    print("\nStep 2: Performing K-means clustering...")
    print("-"*40)
    
    # Determine number of clusters (using first genres)
    first_genres = [re.search(r'[\w\-]+', g).group(0) for g in movies['genres'].dropna() if re.search(r'[\w\-]+', g)]
    n_clusters = len(set(first_genres))
    print(f"Using {n_clusters} clusters (number of unique first genres)")
    
    # Perform clustering
    kmeans_titles = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    clusters_titles = kmeans_titles.fit_predict(X_titles)
    print(f"✓ Title clustering complete")
    
    kmeans_tags = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    clusters_tags = kmeans_tags.fit_predict(X_tags)
    print(f"✓ Tag clustering complete")
    
    kmeans_genre = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    clusters_genre = kmeans_genre.fit_predict(X_genre)
    print(f"✓ Genre clustering complete")
    
    # Step 3: Three-way comparison visualization
    print("\nStep 3: Creating three-way comparison visualization...")
    print("-"*40)
    separation_scores = three_way_comparison_visualization(
        X_titles, X_tags, X_genre,
        title_df, tag_df, genre_df,
        n_sample=500
    )
    
    # Step 4: Comprehensive clustering analysis
    print("\nStep 4: Running comprehensive clustering analysis...")
    print("-"*40)
    metrics_df = comprehensive_clustering_analysis(
        X_titles, X_tags, X_genre,
        clusters_titles, clusters_tags, clusters_genre,
        title_df, tag_df, genre_df
    )
    
    # Step 5: Summary and conclusions
    print("\n" + "="*80)
    print("SUMMARY OF RESULTS")
    print("="*80)
    
    print("\nSeparation Scores (higher is better):")
    for method, score in separation_scores.items():
        print(f"  {method}: {score:.3f}")
    
    print("\nClustering Quality Metrics:")
    print(metrics_df.to_string(index=False))
    
    print("\n" + "="*80)
    print("CONCLUSIONS")
    print("="*80)
    
    # Determine best method
    best_ari_method = metrics_df.loc[metrics_df['ARI'].idxmax(), 'Method']
    best_nmi_method = metrics_df.loc[metrics_df['NMI'].idxmax(), 'Method']
    
    print(f"\n1. Best method by Adjusted Rand Index: {best_ari_method}")
    print(f"2. Best method by Normalized Mutual Info: {best_nmi_method}")
    print("\n3. Key insights:")
    print("   - Title embeddings capture thematic similarities between movies")
    print("   - Tag embeddings reflect user perceptions and descriptions")
    print("   - Genre embeddings provide a baseline/ground truth comparison")
    print("   - The clustering quality differences show how each representation")
    print("     captures different aspects of movie similarity")
    
    # Return all results
    return {
        'embeddings': {
            'titles': (X_titles, title_df),
            'tags': (X_tags, tag_df),
            'genre': (X_genre, genre_df)
        },
        'clusters': {
            'titles': clusters_titles,
            'tags': clusters_tags,
            'genre': clusters_genre
        },
        'metrics': metrics_df,
        'separation_scores': separation_scores,
        'n_clusters': n_clusters
    }

# ============================================
# EXECUTION CODE
# ============================================

if __name__ == "__main__":
    # Load our data
    print("Loading data...")
    movies = pd.read_csv('movielens_movie.csv')
    user_tags = pd.read_csv('movielens_tag.csv')
    
    # Run the complete analysis
    results = run_complete_analysis(movies, user_tags, Glove, GLOVE_M)
    
```
