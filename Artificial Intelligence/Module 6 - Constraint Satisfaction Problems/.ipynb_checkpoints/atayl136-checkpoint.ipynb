## CSP: Map Coloring

In this programming assignment, you will be using your new understanding of **Constraint Satisfaction Problems** to color maps. As we know from the [Four Color Theorem](http://en.wikipedia.org/wiki/Four_color_theorem) any division of a plane into contiguous regions can be colored such that no two adjacent regions are the same color by using at most four colors.

From the book, we know that we can translate this problem into a CSP where the map is represented as a [planar graph](http://en.wikipedia.org/wiki/Planar_graph) and the goal is to color all the nodes such that no adjacent nodes are colored the same color.

As with most AI problems, this requires us to figure out how best to represent the problem--and the solution--given the high and low level data structures and types at our disposal. For this problem, we'll settle on a Dict which contains at least two keys: "nodes" which is an *ordered* List of Strings that represents each node or vertex in the planar graph and "edges" which contains a List of Tuples that represent edges between nodes. The Tuples are of ints that represent the index of the node in the "nodes" list.

````python
def draw_map(name, planar_map, size, color_assignments=None):
    def as_dictionary(a_list):
        dct = {}
        for i, e in enumerate(a_list):
            dct[i] = e
        return dct
    
    G = nx.Graph()
    
    labels = as_dictionary(planar_map["nodes"])
    pos = as_dictionary(planar_map["coordinates"])
    
    # create a List of Nodes as indices to match the "edges" entry.
    nodes = [n for n in range(0, len(planar_map["nodes"]))]

    if color_assignments:
        colors = [c for n, c in color_assignments]
    else:
        colors = ['red' for c in range(0,len(planar_map[ "nodes"]))]

    G.add_nodes_from(nodes)
    G.add_edges_from(planar_map[ "edges"])

    plt.figure(figsize=size, dpi=100)
    plt.title(name)
    nx.draw(G, node_color = colors, with_labels = True, labels = labels, pos = pos)
    
#    plt.savefig(name + ".png")
````

## select_unassigned_variable documentation

Selects an unassigned variable (node) to assign next, using the Degree Heuristic.  

The Degree Heuristic selects the variable with the highest degree (i.e., the  
most unassigned neighbors), under the assumption that it is the most constrained  
and likely to cause a failure sooner.  

Args:  
    assignment (dict): Current variable assignments {node_name: color}.  
    nodes (list): List of all node names.  
    neighbors (dict): Dictionary mapping node names to lists of neighboring nodes.  
    domains (dict): Current domains {node_name: [available_colors]}.  
    trace (bool): If True, outputs tracing information.  

Returns:  
    str: The selected unassigned variable (node name).  

````python
def select_unassigned_variable(assignment, nodes, neighbors, domains, trace):
    unassigned_vars = [v for v in nodes if v not in assignment]
    # Degree heuristic: select variable with highest degree (most neighbors)
    degrees = {v: len(neighbors[v]) for v in unassigned_vars}
    max_degree = max(degrees.values())
    candidates = [v for v in degrees if degrees[v] == max_degree]
    # Break ties by node name (or MRV if implemented)
    selected_var = sorted(candidates)[0]
    if trace:
        print(f"Selecting variable {selected_var} using degree heuristic.")
    return selected_var

````

## lcv__key documentation  (nested function)

        Helper function to compute the heuristic value for sorting.
        Counts how many values in neighboring domains would be eliminated
        if 'var' is assigned 'value'.

````python
def lcv_key(var, value, assignment, domains, neighbors):
    count = 0
    for neighbor in neighbors[var]:
        if neighbor not in assignment and value in domains[neighbor]:
            count += 1
    return count
````

````python
# lcv_key unit tests


# Test Case 1: Value Eliminates Least Options
nodes = ['A', 'B', 'C']
neighbors = {'A': ['B', 'C'], 'B': ['A'], 'C': ['A']}
assignment = {}
domains = {'A': ['red', 'green'], 'B': ['red'], 'C': ['red', 'green']}

# Expected: 'green' is least constraining for 'A' (eliminates fewer options for 'C')
lcv_counts = {val: lcv_key('A', val, assignment, domains, neighbors) for val in domains['A']}
print("Test Case 1 - LCV Counts:", lcv_counts)
assert lcv_counts['green'] < lcv_counts['red']
print("Passed Test Case 1\n")


# Test Case 2: All Values Equally Constraining
domains = {'A': ['red', 'green'], 'B': ['blue'], 'C': ['blue']}
# Neither 'red' nor 'green' eliminate any options for neighbors

lcv_counts = {val: lcv_key('A', val, assignment, domains, neighbors) for val in domains['A']}
print("Test Case 2 - LCV Counts:", lcv_counts)
assert lcv_counts['red'] == lcv_counts['green'] == 0
print("Passed Test Case 2\n")


# Test Case 3: Value Eliminates All Options for Neighbor
domains = {'A': ['red', 'green'], 'B': ['red'], 'C': ['green']}
# Assigning 'red' to 'A' would eliminate all options for 'B'

lcv_counts = {val: lcv_key('A', val, assignment, domains, neighbors) for val in domains['A']}
print("Test Case 3 - LCV Counts:", lcv_counts)
assert lcv_counts['red'] == 1  # Eliminates 'red' from 'B'
assert lcv_counts['green'] == 1  # Eliminates 'green' from 'C'
print("Passed Test Case 3\n")
````

## order-domain_values documentation

Orders the domain values for a variable using the Least Constraining Value heuristic.  

The Least Constraining Value heuristic prefers the value that rules out the fewest  
choices for the neighboring unassigned variables.  

Args:  
    var (str): The variable (node name) to order domain values for.    
    assignment (dict): Current variable assignments {node_name: color}.  
    domains (dict): Current domains {node_name: [available_colors]}.  
    neighbors (dict): Dictionary mapping node names to lists of neighboring nodes.  
    trace (bool): If True, outputs tracing information.  

Returns:  
    list: A list of colors (values) ordered according to the Least Constraining  
          Value heuristic.  


        

````python

def order_domain_values(var, assignment, domains, neighbors, trace):
    ordered_values = sorted(domains[var], key=lambda val: lcv_key(var, val, assignment, domains, neighbors))
    if trace:
        print(f"Ordering domain values for variable {var}: {ordered_values}")
    return ordered_values
````

````python
# Test Case 1 - order_domain_values with Least Constraining Value
nodes = ['A', 'B', 'C']
neighbors = {'A': ['B', 'C'], 'B': ['A'], 'C': ['A']}
assignment = {}
domains = {
    'A': ['red', 'green'],
    'B': ['red'],
    'C': ['red', 'green']
}

# Expected: 'green' is less constraining for 'A' because assigning 'red' to 'A' would eliminate 'red' from 'B' and 'C', whereas 'green' only affects 'C'.

ordered_values = order_domain_values('A', assignment, domains, neighbors, trace=True)

print("Ordered Values for 'A':", ordered_values)

# Verify that 'green' comes before 'red' since it is less constraining.
assert ordered_values == ['green', 'red']

print("Passed Test Case 1\n")


#Test Case 2 - order_domain_values with Equal Constraints
nodes = ['A', 'B', 'C']
neighbors = {'A': ['B', 'C'], 'B': ['A'], 'C': ['A']}
assignment = {}
domains = {
    'A': ['red', 'green'],
    'B': ['blue'],
    'C': ['yellow']
}

# Expected: Since 'A's assignment doesn't affect 'B' or 'C', all values are equally constraining.

ordered_values = order_domain_values('A', assignment, domains, neighbors, trace=True)

print("Ordered Values for 'A':", ordered_values)

# Since all values are equally constraining, the original order should be maintained.
assert ordered_values == ['red', 'green']

print("Passed Test Case 2\n")


# Test Case 3 - order_domain_values where one value eliminates neighbor's domain
nodes = ['A', 'B']
neighbors = {'A': ['B'], 'B': ['A']}
assignment = {}
domains = {
    'A': ['red', 'green'],
    'B': ['red']
}

# Expected: Assigning 'red' to 'A' would eliminate 'red' from 'B', leaving it with no options. So 'green' is less constraining.

ordered_values = order_domain_values('A', assignment, domains, neighbors, trace=True)

print("Ordered Values for 'A':", ordered_values)

# Verify that 'green' comes before 'red'
assert ordered_values == ['green', 'red']

print("Passed Test Case 3\n")
````

## forward_check documentation

Performs forward checking after assigning a value to a variable.

Forward checking removes the assigned value from the domains of unassigned
neighboring variables. If any neighbor ends up with an empty domain, it
indicates an inconsistency, and the function returns False.

Args:
    var (str): The variable (node name) that has been assigned a value.
    value (str): The value (color) assigned to 'var'.
    assignment (dict): Current variable assignments {node_name: color}.
    domains (dict): Current domains {node_name: [available_colors]}.
    neighbors (dict): Dictionary mapping node names to lists of neighboring nodes.

Returns:
    tuple:
        - bool: True if assignment is still consistent after forward checking,
                False if an inconsistency is detected.
        - list of tuples: A list of (neighbor, removed_value) pairs representing
                          the values removed from neighbors' domains.


````python
def forward_check(var, value, assignment, domains, neighbors):
    removals = []
    for neighbor in neighbors[var]:
        if neighbor not in assignment and value in domains[neighbor]:
            domains[neighbor].remove(value)
            removals.append((neighbor, value))
            if not domains[neighbor]:
                return False, removals  # Domain wiped out
    return True, removals
````

````python
# forward_check unit tests

# Test Case 1: No Inconsistency
assignment = {}
domains = {'A': ['red', 'green'], 'B': ['red'], 'C': ['green']}
value = 'red'
var = 'A'

consistent, removals = forward_check(var, value, assignment, domains, neighbors)
print("Test Case 1 - Consistent Assignment:")
print("Consistent:", consistent)
print("Removals:", removals)

assert removals == [('B', 'red')]
print("Passed Test Case 1\n")


# Test Case 2: Inconsistency Detected 
nodes = ['A', 'B']
neighbors = {'A': ['B'], 'B': ['A']}
assignment = {}  # No variables assigned yet
domains = {
    'A': ['red', 'green'],
    'B': ['red']
}
var = 'A'
value = 'red'

# Perform forward checking
consistent, removals = forward_check(var, value, assignment, domains, neighbors)

print("Consistent:", consistent)
print("Removals:", removals)

# Expected: Consistent should be False, as 'B' will have no values left after 'red' is removed
assert consistent is False

print("Passed Test Case 2\n")



# Test Case 3: Multiple Removals Without Inconsistency
nodes = ['A', 'B', 'C']
neighbors = {'A': ['B', 'C'], 'B': ['A'], 'C': ['A']}
assignment = {}  # No variables assigned yet
domains = {
    'A': ['red', 'green'],
    'B': ['red', 'green'],
    'C': ['red', 'green']
}
var = 'A'
value = 'red'

# Perform forward checking
consistent, removals = forward_check(var, value, assignment, domains, neighbors)

print("Consistent:", consistent)
print("Removals:", removals)

# Expected: Both 'B' and 'C' will have 'red' removed but still have 'green'
# So consistent should be True.

# Corrected assertion using set comparison
assert consistent is True
assert set(removals) == {('B', 'red'), ('C', 'red')}

print("Passed Test Case 3\n")
````

## restore_domains documentation  

Restores the domains of variables by adding back the values that were removed  
during forward checking.  

This function is called during backtracking to undo the effects of forward checking  
when a variable assignment leads to a dead end.  

Args:  
    removals (list of tuples): A list of (variable, value) pairs representing  
                               values removed from domains.  
    domains (dict): Current domains {node_name: [available_colors]}.  

````python
def restore_domains(removals, domains):
    for var, value in removals:
        domains[var].append(value)
````

````python
# unit tests

# Test Case 1: Restore Single Removal
domains = {'B': ['green'], 'C': ['red', 'green']}
removals = [('B', 'red')]

restore_domains(removals, domains)
print("Test Case 1 - Domains After Restoration:")
print("Domains:", domains)
assert domains['B'] == ['green', 'red']
print("Passed Test Case 1\n")


# Test Case 2: Restore Multiple Removals
domains = {'B': ['green'], 'C': ['green']}
removals = [('B', 'red'), ('C', 'red')]

restore_domains(removals, domains)
print("Test Case 2 - Domains After Restoration:")
print("Domains:", domains)
assert domains['B'] == ['green', 'red']
assert domains['C'] == ['green', 'red']
print("Passed Test Case 2\n")


# Test Case 3: Restore When Domains are Empty
domains = {'B': [], 'C': ['green']}
removals = [('B', 'red')]

restore_domains(removals, domains)
print("Test Case 3 - Domains After Restoration:")
print("Domains:", domains)
assert domains['B'] == ['red']
print("Passed Test Case 3\n")

````

## backtrack documentation  

Recursively attempts to find a valid color assignment for the map using backtracking  
search with forward checking.  

This function implements the core CSP solver logic, selecting variables and values  
based on heuristics and recursively exploring the search space.  

Args:  
    assignment (dict): Current variable assignments {node_name: c olor}.  
    domains (dict): Current domains {node_name: [available_colors]}.  
    nodes (list): List of all node names.  
    neighbors (dict): Dictionary mapping node names to lists of neighboring nodes.  
    color_list (list): List of available colors.  
    trace (bool): If True, outputs tracing information.  

Returns:  
    bool: True if a complete and consistent assignment is found, False otherwise.  


````python
def backtrack(assignment, domains, nodes, neighbors, color_list, trace):
    if len(assignment) == len(nodes):
        return True
    var = select_unassigned_variable(assignment, nodes, neighbors, domains, trace)
    for value in order_domain_values(var, assignment, domains, neighbors, trace):
        if trace:
            print(f"Trying value '{value}' for variable {var}.")
        assignment[var] = value
        consistent, removals = forward_check(var, value, assignment, domains, neighbors)
        if consistent:
            result = backtrack(assignment, domains, nodes, neighbors, color_list, trace)
            if result:
                return True
        if trace and not consistent:
            print(f"Backtracking from variable {var}, value '{value}' caused inconsistency.")
        del assignment[var]
        restore_domains(removals, domains)
    return False
````

````python
#  unit tests


# Test Case 1: Simple Map with Solution
assignment = {}
domains = {'A': ['red', 'green'], 'B': ['red', 'green']}
nodes = ['A', 'B']
neighbors = {'A': ['B'], 'B': ['A']}
color_list = ['red', 'green']
trace = True

result = backtrack(assignment, domains, nodes, neighbors, color_list, trace)
print("Test Case 1 - Solution Found:")
print("Assignment:", assignment)
assert result is True
assert assignment['A'] != assignment['B']
print("Passed Test Case 1\n")


# Test Case 2: Map with No Solution
assignment = {}
domains = {'A': ['red'], 'B': ['red']}
# Both nodes can only be 'red' and are connected

result = backtrack(assignment, domains, nodes, neighbors, color_list, trace)
print("Test Case 2 - No Solution:")
print("Assignment:", assignment)
assert result is False
print("Passed Test Case 2\n")


# Test Case 3: Larger Map
assignment = {}
domains = {'A': ['red', 'green'], 'B': ['red', 'green'], 'C': ['red', 'green']}
nodes = ['A', 'B', 'C']
neighbors = {'A': ['B', 'C'], 'B': ['A', 'C'], 'C': ['A', 'B']}
color_list = ['red', 'green']

result = backtrack(assignment, domains, nodes, neighbors, color_list, trace)
print("Test Case 3 - Larger Map with No Solution:")
print("Assignment:", assignment)
assert result is False
print("Passed Test Case 3\n")
````

# Color Map Documentation  


Attempts to color the given planar map using the specified list of colors,  
ensuring that no two adjacent nodes have the same color.  

This function implements a Constraint Satisfaction Problem (CSP) solver  
using backtracking search with forward checking. It employs the Degree    
Heuristic for variable selection and the Least Constraining Value heuristic  
for value ordering.  

Args:  
    planar_map (dict): A dictionary representing the map, containing:  
        - 'nodes': List of node names (strings).  
        - 'edges': List of tuples representing edges between nodes.  
    color_list (list): A list of color names (strings) to use for coloring.  
    trace (bool, optional): If True, enables tracing of the algorithm's steps.  
                            Defaults to False.  

Returns:
    list of tuples: A list of (node_name, color) pairs representing the color  
                    assignment, in the same order as the nodes in the planar_map.  
    None: If no valid coloring is found.  


````python
def color_map(planar_map, color_list, trace=False):
    # Extract nodes and edges
    nodes = planar_map['nodes']
    edges = planar_map['edges']
    neighbors = {node: [] for node in nodes}
    
    # Build the neighbors map
    for edge in edges:
        n1 = nodes[edge[0]]
        n2 = nodes[edge[1]]
        neighbors[n1].append(n2)
        neighbors[n2].append(n1)
    
    # Initialize domains for each node
    domains = {node: list(color_list) for node in nodes}
    assignment = {}
    
    # Start the backtracking algorithm
    if backtrack(assignment, domains, nodes, neighbors, color_list, trace):
        # Return assignments in the original node order
        return [(node, assignment[node]) for node in nodes]
    else:
        return None
````

````python
# color_map unit tests

# Test Case 1: Simple map with two connected nodes
simple_map = {
    'nodes': ['A', 'B'],
    'edges': [(0, 1)],
    'coordinates': [(0, 0), (1, 1)]
}
colors = ['red', 'blue']

result = color_map(simple_map, colors)
print("Test Case 1 - Simple Map with Valid Coloring:")
print("Result:", result)
assert result == [('A', 'red'), ('B', 'blue')] or result == [('A', 'blue'), ('B', 'red')]
print("Passed Test Case 1\n")


# Test Case 2: Map with a triangle (needs at least 3 colors)
triangle_map = {
    'nodes': ['A', 'B', 'C'],
    'edges': [(0, 1), (1, 2), (2, 0)],
    'coordinates': [(0, 0), (1, 1), (2, 0)]
}
colors = ['red', 'blue']

result = color_map(triangle_map, colors)
print("Test Case 2 - Triangle Map with Insufficient Colors:")
print("Result:", result)
assert result is None
print("Passed Test Case 2\n")


# Test Case 3: Map with four nodes forming a square
square_map = {
    'nodes': ['A', 'B', 'C', 'D'],
    'edges': [(0, 1), (1, 2), (2, 3), (3, 0)],
    'coordinates': [(0, 0), (1, 0), (1, 1), (0, 1)]
}
colors = ['red', 'blue']

result = color_map(square_map, colors)
print("Test Case 3 - Square Map with Valid Coloring:")
print("Result:", result)
assert result is not None
assigned_colors = [color for node, color in result]
assert len(set(assigned_colors)) <= 2  # Should use only 2 colors
print("Passed Test Case 3\n")
````
