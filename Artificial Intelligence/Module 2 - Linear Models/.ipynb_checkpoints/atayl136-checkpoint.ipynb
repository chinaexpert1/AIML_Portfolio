The data that comes in, however, is noisy. The values are never exactly 0 and 1. In order to mimic this we need a `blur` function.

We will assume that noise is normally distributed. For values that should be 0, the noisy values are distributed $N(0.10, 0.05)$. For values that should be 1, the noisy values are distributed $N(0.9, 0.10)$.

````python
def blur( data):
    def apply_noise( value):
        if value < 0.5:
            v = random.gauss(0.30, 0.07) # (0.10, 0.05)
            if v < 0.0:
                return 0.0
            if v > 0.75:
                return 0.75
            return v
        else:
            v = random.gauss(0.70, 0.07) # (0.90, 0.10)
            if v < 0.25:
                return 0.25
            if v > 1.00:
                return 1.00
            return v
    noisy_readings = [apply_noise( v) for v in data[0:-1]]
    return noisy_readings + [data[-1]]
````

### sigmoid documentation
**Description**:  
This function computes the sigmoid (logistic) function on the input `z`. The sigmoid function is essential for logistic regression as it converts linear combinations of features into probabilities between 0 and 1. These probabilities are used to classify observations into binary classes.

**Parameters**:
- `z` (numpy array): The input value(s) to apply the sigmoid function on.

**Return**:
- (numpy array): The sigmoid of the input, a value between 0 and 1 for each input element. It transforms the (weighted) features./

**Use**:
- Used by the `learn_model`and `apply_model` functions to compute prediction probabilities.


````python
# Helper Function: Sigmoid

import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
````

### feature_scaling documentation
**Description**:  
This function standardizes the feature set by subtracting the mean and dividing by the standard deviation of each feature. This process ensures that all features have the same scale, which helps in improving the convergence speed of the logistic regression model. I had to look up the math definition to remind myself what it was I actually forgot. I was thinking min max scaling would work?

**Parameters**:
- `X` (numpy array): The feature matrix where each row is an observation and each column is a feature.

**Return**:
- (numpy array): The scaled feature matrix where each feature has mean 0 and standard deviation 1.

**Use**:
- Used by the `learn_model` and `apply_model` functions for feature preprocessing of the train and test set, respectively.


````python
# Helper function: feature scaling

def feature_scaling(X):
    # standardize features to have mean = 0 and variance = 1.
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    return (X - mean) / (std + 1e-5)  
````

## Main Functions

Use `generate_data` to generate a balanced set of blurred "hills" and "not-hills" examples to test that the function is working.

````python
def generate_data( data, n, key_label):
    labels = list(data.keys())
    labels.remove(key_label)

    total_labels = len(labels)
    result = []
    # create n "not label" and code as y=0
    count = 1
    while count <= n:
        label = labels[count % total_labels]
        datum = blur(random.choice(data[label]))
        xs = datum[0:-1]
        result.append((xs, 0))
        count += 1

    # create n "label" and code as y=1
    for _ in range(n):
        datum = blur(random.choice(data[key_label]))
        xs = datum[0:-1]
        result.append((xs, 1))
    random.shuffle(result)
    return result
````


### learn_model documentation
**Description**:  
This function implements logistic regression using gradient descent. It learns the model parameters (the weights) `Theta` that minimize the error between the predicted and actual labels. Regularization was applied to prevent overfitting. This function is the core of the algorithm, responsible for training the model.

**Parameters**:
- `train_data` (list of tuples): The training dataset, where each element is a tuple containing a feature vector and its corresponding target label.
- `reg_lambda` (float): The regularization term to control overfitting.
- `alpha` (float): The learning rate that controls the step size in gradient descent.
- `verbose` (bool): If `True`, prints the error at every 100 iterations.

**Return**:
- `Theta` (numpy array): The learned model weights after gradient descent converges.

**USe**:
- Uses the `sigmoid` function to compute prediction probabilities.
- Used by the overall pipeline to train the logistic regression model.


````python
def learn_model(train_data, reg_lambda=0.05, alpha=0.01, verbose=True):
    # preprocess the data
    X_train = np.array([data[0] for data in train_data], dtype=float)
    y_train = np.array([data[1] for data in train_data], dtype=float)

    X_train = feature_scaling(X_train)

    m, n = X_train.shape
    Theta = np.random.randn(n) * 0.01  # initialize weights
    epsilon = 1e-5  # convergence threshold
    max_iter = 10000
    min_iter = 100  # minimum iterations
    error_prev = float('inf')

    for i in range(max_iter):
        # prediction
        z = np.dot(X_train, Theta)
        h = sigmoid(z)

        # compute gradient and update Theta
        gradient = np.dot(X_train.T, (h - y_train)) / m
        Theta -= alpha * (gradient + reg_lambda * Theta)  # applied L2 regularization

        # compute the error (log-loss)
        error = - (1 / m) * (np.dot(y_train, np.log(h + epsilon)) + np.dot(1 - y_train, np.log(1 - h + epsilon)))

        if verbose and i % 100 == 0:
            print(f"Iteration {i}, Error: {error}")

        error_prev = error

    print(f"Training complete after {i+1} iterations!")
    return Theta
````


### apply_model
**Description**:  
This function applies the learned logistic regression model to test data, producing predictions based on a given decision threshold. The predictions are probabilities converted to binary classifications (0 or 1). This function is critical for testing the model's performance on unseen data.

**Formal Parameters**:
- `model` (numpy array): The learned model parameters (weights).
- `test_data` (list of tuples): The test dataset, where each element is a tuple containing a feature vector and its corresponding label.
- `threshold` (float): The probability threshold used to classify predictions as positive (1) or negative (0).

**Return**:
- `results` (list of tuples): A list of tuples where each tuple contains the actual label and the predicted label.

**Linkage**:
- Uses the `sigmoid` function to compute prediction probabilities.
- Uses the `feature_scaling` function for preprocessing features.
- The results are passed to the `evaluate` function for performance evaluation.


````python
def apply_model(model, test_data, threshold=0.7):  # Adjust threshold here
    # preprocess the data
    X_test = np.array([data[0] for data in test_data], dtype=float)
    y_test = np.array([data[1] for data in test_data], dtype=float)

    # feature scaling
    X_test = feature_scaling(X_test)

    # prediction using sigmoid function
    predictions = sigmoid(np.dot(X_test, model))

    # apply the adjusted threshold to convert probabilities to predictions
    return [(y_test[i], 1 if pred >= threshold else 0) for i, pred in enumerate(predictions)]
````


### evaluate documentaation
**Description**:  
This function evaluates the performance of the logistic regression model by calculating the error rate and constructing a confusion matrix. It reports how well the model is classifying observations into the correct classes, and helps identify issues like false positives and false negatives.

**Formal Parameters**:
- `results` (list of tuples): The predicted results, where each tuple contains the actual label and the predicted label.

**Return**:
- None (prints the error rate and confusion matrix).

**Linkage**:
- Receives results from the `apply_model`function for evaluation.


````python
def evaluate(results):
    # calculate error rate
    error_rate = np.mean([1 if np.round(pred) != actual else 0 for actual, pred in results])
    print(f"Error rate: {error_rate * 100:.2f}%")
    
    # confusion matrix
    tp = sum(1 for actual, pred in results if actual == 1 and np.round(pred) == 1)
    tn = sum(1 for actual, pred in results if actual == 0 and np.round(pred) == 0)
    fp = sum(1 for actual, pred in results if actual == 0 and np.round(pred) == 1)
    fn = sum(1 for actual, pred in results if actual == 1 and np.round(pred) == 0)

    print("Confusion Matrix:")
    print(f"TP: {tp}, FP: {fp}, TN: {tn}, FN: {fn}")
````

````python
# evaluate function unit tests

# Test 1: ensure error rate is 1 when all predictionss are correct
results_test = [(1, 1), (0, 0), (1, 1), (0, 0)]
try:
    assert np.isclose(np.mean([1 if pred != actual else 0 for actual, pred in results_test]), 0)
    print("evaluate Test 1 Passed: Error rate is 0 when all predictions are correct")
except:
    print("evaluate Test 1 was failed.")

# test 2: ensure confusion matrix counts correct TP and TN
try:
    evaluate(results_test)  # should print TP: 2, FP: 0, TN: 2, FN: 0
    print("evaluate Test 2 Passed: Confusion matrix correctly counts TP and TN")
except:
    print("evaluate Test 2 was failed.")

# Test 3: ensure confusion matrix counts correct FP and FN
results_test = [(1, 0), (0, 1), (1, 0), (0, 1)]
try:
    evaluate(results_test)  # should print TP: 0, FP: 2, TN: 0, FN: 2
    print("evaluate Test 3 Passed: Confusion matrix correctly counts FP and FN")
except:
    print("evaluate Test 3 was failed.")
````
