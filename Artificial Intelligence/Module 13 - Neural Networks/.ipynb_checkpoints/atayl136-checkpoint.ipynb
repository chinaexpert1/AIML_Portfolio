The data that comes in, however, is noisy. The values are never exactly 0 and 1. In order to mimic this we need a `blur` function.

We will assume that noise is normally distributed. For values that should be 0, the noisy values are distributed $N(0.10, 0.05)$. For values should be 1, the noisy values are distributed $N(0.9, 0.10)$.

````python
def blur( data):
    def apply_noise( value):
        if value < 0.5:
            v = random.gauss( 0.10, 0.05)
            if v < 0.0:
                return 0.0
            if v > 0.75:
                return 0.75
            return v
        else:
            v = random.gauss( 0.90, 0.10)
            if v < 0.25:
                return 0.25
            if v > 1.00:
                return 1.00
            return v
    noisy_readings = [apply_noise( v) for v in data[0:-1]]
    return noisy_readings + [data[-1]]
````

## softmax documentation


Computes the softmax activation function.

The softmax function converts a vector of raw scores (logits) into probabilities that sum to one. It is commonly used in the output layer of a neural network for multi-class classification problems.

Parameters:

z : numpy.ndarray
    The input array of logits from the neural network.

Returns:

numpy.ndarray
    An array of probabilities corresponding to each class, where each probability is between 0 and 1.


````python
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # For numerical stability
    return exp_z / np.sum(exp_z)
````

````python

print("Testing softmax function...")


z = np.array([0.0, 0.0, 0.0])
softmax_output = softmax(z)
assert isinstance(softmax_output, np.ndarray), "Output should be a NumPy array"
assert np.allclose(np.sum(softmax_output), 1.0), "Probabilities should sum to 1"
assert np.allclose(softmax_output, [1/3, 1/3, 1/3]), "Softmax of zeros should be a uniform distribution"

z = np.array([1.0, 2.0, 3.0])
softmax_output = softmax(z)
expected_output = np.exp(z) / np.sum(np.exp(z))
assert np.allclose(softmax_output, expected_output), "Softmax output is incorrect"

print("softmax function passed the 4 tests.\n")
````

## sigmoid documentation


Computes the sigmoid activation function.

The sigmoid function is defined as 1 / (1 + e^(-z)). It maps any number into the range (0, 1).

Parameters:

z : float or numpy.ndarray
    The input value or array of values for which to compute the sigmoid function.

Returns:

float or numpy.ndarray
    The sigmoid of the input value(s), having the same shape as the input.


````python
# Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
````

````python

print("Testing sigmoid function...")

z = 0.0
sigmoid_output = sigmoid(z)
assert isinstance(sigmoid_output, float), "Output should be a float"
assert np.isclose(sigmoid_output, 0.5), "Sigmoid of 0 should be 0.5"

z = -2.0
sigmoid_output = sigmoid(z)
expected_output = 1 / (1 + np.exp(-z))
assert np.isclose(sigmoid_output, expected_output), "Sigmoid output is incorrect"

print("sigmoid function passed the 3 tests.\n")
````

## sigmoid_derivative


Computes the derivative of the sigmoid function.

The derivative of the sigmoid function is use during the backpropagation step in neural networks. It is calculated as sigmoid(z) * (1 - sigmoid(z)).

Parameters:

z : float or numpy.ndarray
    The input value or array of values for which to compute the derivative of the sigmoid function.

Returns:

float or numpy.ndarray
    The derivative of the sigmoid function at the input values, having the same shape as the input.


````python
def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)
````

````python

print("Testing sigmoid_derivative function...")

z = 0.0
sigmoid_derivative_output = sigmoid_derivative(z)
assert isinstance(sigmoid_derivative_output, float), "Output should be a float"
assert np.isclose(sigmoid_derivative_output, 0.25), "Sigmoid derivative at 0 should be 0.25"

z = 2.0
sigmoid_derivative_output = sigmoid_derivative(z)
s = sigmoid(z)
expected_output = s * (1 - s)
assert np.isclose(sigmoid_derivative_output, expected_output), "Sigmoid derivative output is incorrect"

print("sigmoid_derivative function passed the 3 tests.\n")
````

### generate_data

Generates an endless supply of blurred data from a collection of terrain prototypes.

* `data`: Dict[Str, List[Any]] - a Dictionary of "clean" prototypes for each landscape type.
* `n`: Int - the number of blurred examples of each terrain type to return.

returns

* List[List[Any]] - a List of Lists. Each individual List is a blurred example of a terrain type, generated from the prototype.

````python
def generate_data(clean_data, num_examples_per_class):
    label_to_one_hot = {
        'hills':  [1, 0, 0, 0],
        'swamp':  [0, 1, 0, 0],
        'forest': [0, 0, 1, 0],
        'plains': [0, 0, 0, 1]
    }
    data = []
    for terrain, examples in clean_data.items():
        for _ in range(num_examples_per_class):
            clean_example = random.choice(examples)
            noisy_example = blur(clean_example)
            x = noisy_example[:-1]
            y = label_to_one_hot[noisy_example[-1]]
            data.append((x, y))
    return data
````

## learn_model documentation  


Trains a neural network model using the provided training data.  

This function implements a feedforward neural network with one hidden layer.  
It uses backpropagation to update the weights and biases based on the training data.  


Parameters:  

training_data : list of tuples  
    A list where each element is a tuple (x, y):  
    - x : list of floats  
        The input features for a single training example (length 16).  
    - y : list of ints  
        The one-hot encoded label vector for the training example (length 4).  

num_hidden_nodes : int  
    The number of nodes in the hidden layer.  

verbose : bool  
    If True, the function prints out the total loss at specified iterations  
    to monitor the training process.  

Returns:  

model : tuple   
    A tuple containing:  
    - hidden_layer_weights : numpy.ndarray  
        The weights of the hidden layer, including biases.  
        Shape: (num_hidden_nodes, input_size + 1)  
    - output_layer_weights : numpy.ndarray  
        The weights of the output layer, including biases.  
        Shape: (output_size, num_hidden_nodes + 1)  




````python
def learn_model(training_data, num_hidden_nodes, verbose):
    input_size = 16
    hidden_size = num_hidden_nodes
    output_size = 4
    max_iterations = 5000
    initial_alpha = 0.0001
    decay_rate = 0.001



    hidden_layer_weights = np.random.uniform(-0.5, 0.5, size=(hidden_size, input_size + 1))
    output_layer_weights = np.random.uniform(-0.5, 0.5, size=(output_size, hidden_size + 1))

    for iteration in range(1, max_iterations + 1):
        total_loss = 0.0
        alpha = initial_alpha / (1 + decay_rate * iteration)
        random.shuffle(training_data)

        for x, y in training_data:
            x_with_bias = np.array([1.0] + x)
            y = np.array(y)

            # Forward propagation
            z_hidden = hidden_layer_weights @ x_with_bias
            a_hidden = sigmoid(z_hidden)
            a_with_bias = np.concatenate(([1.0], a_hidden))
            z_output = output_layer_weights @ a_with_bias
            o_k = softmax(z_output)

            loss = -np.sum(y * np.log(o_k + 1e-15))
            total_loss += loss

            delta_k = o_k - y  
            delta_j = (output_layer_weights[:, 1:].T @ delta_k) * sigmoid_derivative(z_hidden)

            output_layer_weights -= alpha * np.outer(delta_k, a_with_bias)
            hidden_layer_weights -= alpha * np.outer(delta_j, x_with_bias)

        if verbose and (iteration % 1000 == 0 or iteration == 1):
            print(f"Iteration {iteration}, Training Loss: {total_loss}, Learning Rate: {alpha}")

    model = (hidden_layer_weights, output_layer_weights)
    return model
````

````python
print("Testing learn_model function...")


test_training_data = [
    ([0.0] * 16, [1, 0, 0, 0]), 
    ([1.0] * 16, [0, 0, 0, 1])  
]

test_model = learn_model(test_training_data, num_hidden_nodes=2, verbose=False)

assert isinstance(test_model, tuple), "Model should be a tuple"
print("Passed: Model is a tuple.")
assert len(test_model) == 2, "Model should contain two elements"
print("Passed: Model contains two elements.")
hidden_layer_weights, output_layer_weights = test_model
assert isinstance(hidden_layer_weights, np.ndarray), "Hidden layer weights should be a NumPy array"
print("Passed: Hidden layer weights is a NumPy array.")
assert isinstance(output_layer_weights, np.ndarray), "Output layer weights should be a NumPy array"
print("Passed: Output layer weights is a NumPy array.")
assert hidden_layer_weights.shape == (2, 17), "Hidden layer weights should have shape (2, 17)"
print("Passed: Hidden layer weights has shape (2, 17).")
assert output_layer_weights.shape == (4, 3), "Output layer weights should have shape (4, 3)"
print("Passed: Output layer weights has shape (4, 3).")

print("learn_model function passed the tests.\n")
````

### apply_model

**put documentation here based on the assignment description**

## apply_model documentation

Applies the trained neural network model to the provided data.  

This function performs forward propagation using the trained weights to make predictions.  
It can handle both labeled and unlabeled data.  

Parameters:    
  
model : tuple  
    A tuple containing the trained weights of the neural network:  
    - hidden_layer_weights : numpy.ndarray  
        The weights of the hidden layer, including biases.  
    - output_layer_weights : numpy.ndarray  
        The weights of the output layer, including biases.  
data : list  
    A list of examples to which the model will be applied. Each element can be:  
    - Labeled data: a tuple (x, y):  
        - x : list of floats  
            The input features  for a single example.  
        - y : list of ints  
            The one-hot encoded label vector for the example. 
    - Unlabeled data: x  
        - x : list of floats  
            The input features for a single example.  

Returns:  

results : list  
    A list of results for each example in the data:  
    - For labeled data:  
        - A list of tuples for each class: (actual_value, predicted_value)  
    - For unlabeled data:  
        - A list of tuples for each class: (predicted_value, probability)  
    


````python

def apply_model(model, data):
    hidden_layer_weights, output_layer_weights = model
    results = []
    for example in data:
        if len(example) == 2:
            # Labeled data
            x, y = example
        else:
            # Unlabeled data
            x = example
            y = None
        x_with_bias = np.array([1.0] + x)
        z_hidden = hidden_layer_weights @ x_with_bias
        a_hidden = sigmoid(z_hidden)
        a_with_bias = np.concatenate(([1.0], a_hidden))
        z_output = output_layer_weights @ a_with_bias
        o_k = softmax(z_output)
        predicted_class = np.argmax(o_k)
        inferred_values = [1 if k == predicted_class else 0 for k in range(4)]
        probabilities = o_k.tolist()
        if y is not None:
            # Labeled data
            result = list(zip(y, inferred_values))
        else:
            # Unlabeled data
            result = list(zip(inferred_values, probabilities))
        results.append(result)
    return results
````

````python
print("Testing apply_model function...")

test_model = (
    np.random.uniform(-0.5, 0.5, size=(2, 17)),  # Hidden layer weights (2 nodes, 16 features + 1 bias)
    np.random.uniform(-0.5, 0.5, size=(4, 3))    # Output layer weights (4 classes, 2 hidden nodes + 1 bias)
)

test_data = [
    [0.5] * 16  # Example input features without labels
]

test_results = apply_model(test_model, test_data)

assert isinstance(test_results, list), "Results should be a list"
print("Passed: Results is a list.")
assert len(test_results) == 1, f"Should have one result, got {len(test_results)}"
print("Passed: Results has one element.")
result = test_results[0]
assert isinstance(result, list), "Each result should be a list"
print("Passed: Each result is a list.")
assert len(result) == 4, "Each result should have 4 class predictions"
print("Passed: Each result has 4 class predictions.")
predicted_values, probabilities = zip(*result)
assert all(isinstance(pv, int) for pv in predicted_values), "Predicted values should be integers"
print("Passed: Predicted values are integers.")
assert sum(predicted_values) == 1, "Only one class should be predicted as '1'"
print("Passed: Only one class is predicted as '1'.")
assert all(0.0 <= prob <= 1.0 for prob in probabilities), "Probabilities should be between 0 and 1"
print("Passed: Probabilities are between 0 and 1.")

print("apply_model function passed the tests.\n")
````

````python
hidden_nodes_list = [2, 4, 8]
training_examples_per_class = 500  
test_examples_per_class = 200     


train_data = generate_data(clean_data, training_examples_per_class)
test_data = generate_data(clean_data, test_examples_per_class)

error_rates = []

for num_hidden_nodes in hidden_nodes_list:
    print(f"\nTraining model with {num_hidden_nodes} hidden nodes.")
    
    model = learn_model(train_data, num_hidden_nodes=num_hidden_nodes, verbose=True)
    
    train_results = apply_model(model, train_data)
    train_correct = 0
    for res in train_results:
        actual_values, predicted_values = zip(*res)
        if actual_values == predicted_values:
            train_correct += 1
    train_error_rate = 1 - (train_correct / len(train_results))
    
    test_results = apply_model(model, test_data)
    test_correct = 0
    for res in test_results:
        actual_values, predicted_values = zip(*res)
        if actual_values == predicted_values:
            test_correct += 1
    test_error_rate = 1 - (test_correct / len(test_results))
    
    error_rates.append({
        'hidden_nodes': num_hidden_nodes,
        'train_error_rate': train_error_rate,
        'test_error_rate': test_error_rate
    })
    
    print(f"Training Error Rate: {train_error_rate * 100:.2f}%")
    print(f"Test Error Rate: {test_error_rate * 100:.2f}%")

# Determine the best number of hidden nodes based on test error rate
best_model = min(error_rates, key=lambda x: x['test_error_rate'])
print("\nSummary of Error Rates:")
for er in error_rates:
    print(f"Hidden Nodes: {er['hidden_nodes']}, Training Error Rate: {er['train_error_rate']*100:.5f}%, Test Error Rate: {er['test_error_rate']*100:.5f}%")
print(f"\nBest number of hidden nodes: {best_model['hidden_nodes']} with Test Error Rate: {best_model['test_error_rate']*100:.5f}%")
````
