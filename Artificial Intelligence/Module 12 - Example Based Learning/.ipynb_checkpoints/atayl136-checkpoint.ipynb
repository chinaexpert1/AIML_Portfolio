## Load the Data

The function `parse_data` loads the data from the specified file and returns a List of Lists. The outer List is the data set and each element (List) is a specific observation. Each value of an observation is for a particular measurement. This is what we mean by "tidy" data.

The function also returns the *shuffled* data because the data might have been collected in a particular order that *might* bias training.

```python
import random
from typing import List, Dict, Tuple, Callable
```

```python
def parse_data(file_name: str) -> List[List]:
    data = []
    file = open(file_name, "r")
    for line in file:
        datum = [float(value) for value in line.rstrip().split(",")]
        data.append(datum)
    random.shuffle(data)
    return data
```

## Train/Test Splits - n folds

With n fold cross validation, we divide our data set into n subgroups called "folds" and then use those folds for training and testing. You pick n based on the size of your data set. If you have a small data set--100 observations--and you used n=10, each fold would only have 10 observations. That's probably too small. You want at least 30. At the other extreme, we generally don't use n > 10.

With 1,030 observations, n = 10 is fine so we will have 10 folds.
`create_folds` will take a list (xs) and split it into `n` equal folds with each fold containing one-tenth of the observations.

```python
def create_folds(data: List[List[float]], n: int, seed: int = 42) -> List[List[List[float]]]:
    random.seed(seed)
    data.sort(key=lambda x: x[-1])  # Sort by target variable
    folds = [[] for _ in range(n)]
    for i, point in enumerate(data):
        folds[i % n].append(point)
    return folds
```

```python
folds = create_folds(data, 10)
```

```python
len(folds)
```

We always use one of the n folds as a test set (and, sometimes, one of the folds as a *pruning* set but not for kNN), and the remaining folds as a training set.
We need a function that'll take our n folds and return the train and test sets:

```python
def create_train_test(folds: List[List[List]], index: int) -> Tuple[List[List], List[List]]:
    training = []
    test = []
    for i, fold in enumerate(folds):
        if i == index:
            test = fold
        else:
            training = training + fold
    return training, test
```

We can test the function to give us a train and test datasets where the test set is the fold at index 0:

## euclidean_distance documentation

Description: Computes the Euclidean distance between two points in the feature space, ignoring the target value.  
Parameters:  
point1 (List[float]): The first point (features only).  
point2 (List[float]): The second point (features only).  
Returns:  
The Euclidean distance (float) between the two points.  

```python
# Euclidean distance calculation
def euclidean_distance(point1: List[float], point2: List[float]) -> float:
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(point1[:-1], point2[:-1])))
```

## knn_regression documentation  

Description: Implements the k-Nearest Neighbors algorithm for regression, predicting the target value for a query point based on the average target values of the k-nearest neighbors.  

Parameters: 
train_data (List[List[float]]): Training dataset with features and target values.  
query (List[float]): The query point (features only).  
k (int): Number of neighbors to consider.  

Returns:  
Predicted target value (float) for the query point.

```python
# kNN function for regression
def knn_regression(train_data: List[List[float]], query: List[float], k: int = 9) -> float:
    # Calculate distances
    distances = [(euclidean_distance(query, train), train[-1]) for train in train_data]
    # Sort by distance
    distances.sort(key=lambda x: x[0])
    # Select k-nearest neighbors
    nearest_neighbors = distances[:k]
    # Return the mean of the neighbors target values
    return sum(neighbor[1] for neighbor in nearest_neighbors) / k
```

## mean_squared_error_manual documentation

Description: Calculates the Mean Squared Error (MSE) between true and predicted target values.  

Parameters:  
true_values (List[float]): List of true target values.  
predicted_values (List[float]): List of predicted target values.  

Returns:  
Mean Squared Error (float).



```python
# Define a function to calculate Mean Squared Error (MSE)
def mean_squared_error_manual(true_values: List[float], predicted_values: List[float]) -> float:
    errors = [(true - pred) ** 2 for true, pred in zip(true_values, predicted_values)]
    return sum(errors) / len(errors)

# Define train_targets explicitly based on the train data
train_targets = [sample[-1] for sample in train_data]

# Redefine baseline predictions
baseline_prediction = sum(train_targets) / len(train_targets)
baseline_predictions = [baseline_prediction] * len(test_targets)

# Recalculate MSE using the manual function
knn_mse_manual = mean_squared_error_manual(test_targets, predictions)
baseline_mse_manual = mean_squared_error_manual(test_targets, baseline_predictions)

# Display the MSE values
knn_mse_manual, baseline_mse_manual
```

## manual_scale_data documentation

Scales the feature columns of the dataset manually (mean = 0, std = 1).  

Parameters:  
    data (List[List[float]]): Dataset with features and target values.  

Returns:  
    List[List[float]]: Scaled dataset with original target values preserved.  


```python
# Manual implementation of feature scaling
def manual_scale_data(data: List[List[float]]) -> List[List[float]]:
    features = [row[:-1] for row in data]  # Extract features
    targets = [row[-1] for row in data]    # Extract targets

    # Calculate mean and standard deviation for each feature
    means = [sum(col) / len(col) for col in zip(*features)]
    std_devs = [((sum((x - mean) ** 2 for x in col) / len(col)) ** 0.5) for col, mean in zip(zip(*features), means)]

    # Scale features manually
    scaled_features = [
        [(value - mean) / std if std != 0 else 0 for value, mean, std in zip(row, means, std_devs)]
        for row in features
    ]

    return [list(scaled_row) + [target] for scaled_row, target in zip(scaled_features, targets)]


```

```python
# Testing: manual_scale_data
print("\nTesting: manual_scale_data")
test_data = [[10.0, 20.0, 30.0], [40.0, 50.0, 60.0], [70.0, 80.0, 90.0]]
scaled_data = manual_scale_data(test_data)

# Test 1: Check if output is a list of lists
assert isinstance(scaled_data, list) and all(isinstance(row, list) for row in scaled_data), \
    "Test Failed: manual_scale_data does not return a list of lists"
print("Test Passed: manual_scale_data returns a list of lists")

# Test 2: Check if the means of features are approximately 0
means = [sum(feature) / len(feature) for feature in zip(*[row[:-1] for row in scaled_data])]
assert all(abs(mean) < 1e-6 for mean in means), \
    "Test Failed: manual_scale_data does not standardize means to 0"
print("Test Passed: manual_scale_data standardizes means to 0")

# Test 3: Check if the standard deviations of features are approximately 1
std_devs = [
    (sum((x - mean) ** 2 for x in feature) / len(feature)) ** 0.5
    for feature, mean in zip(zip(*[row[:-1] for row in scaled_data]), means)
]
assert all(abs(std_dev - 1) < 1e-6 for std_dev in std_devs), \
    "Test Failed: manual_scale_data does not standardize std deviations to 1"
print("Test Passed: manual_scale_data standardizes std deviations to 1")
```

## k_fold_cross_validation documentation  

Description: Performs k-fold cross-validation for multiple values of k (number of neighbors) in kNN regression and computes the average MSE for each k.  

Parameters:  
data (List[List[float]]): Dataset with features and target values.  
k_values (List[int]): List of k-values to test.  
folds (int): Number of cross-validation folds.  

Returns:  
A list of tuples, where each tuple contains k and the corresponding average MSE.

```python
def k_fold_cross_validation(data: List[List[float]], k_values: List[int], folds: int = 10) -> List[Tuple[int, float]]:
    # Create stratified folds
    data_folds = create_folds(data, folds)
    results = []

    for k in k_values:
        mse_scores = []

        for fold_index in range(folds):
            train, test = create_train_test(data_folds, fold_index)
            test_targets = [sample[-1] for sample in test]
            predictions = [knn_regression(train, sample, k) for sample in test]
            mse = mean_squared_error_manual(test_targets, predictions)
            mse_scores.append(mse)

        avg_mse = sum(mse_scores) / len(mse_scores)
        results.append((k, avg_mse))

    return results

```

```python
# Testing: k_fold_cross_validation
print("\nTesting: k_fold_cross_validation")
test_data = [[i, i * 2, i * 3] for i in range(1, 11)]  # Create a dataset
k_values = [1, 2, 3]
folds = 2
results = k_fold_cross_validation(test_data, k_values, folds=folds)

# Test 1: Check if the output is a list of tuples
assert isinstance(results, list) and all(isinstance(result, tuple) for result in results), \
    "Test Failed: k_fold_cross_validation does not return a list of tuples"
print("Test Passed: k_fold_cross_validation returns a list of tuples")

# Test 2: Check if the number of results matches k_values
assert len(results) == len(k_values), \
    "Test Failed: k_fold_cross_validation does not return results for all k values"
print("Test Passed: k_fold_cross_validation returns results for all k values")

# Test 3: Check if MSE values are floats
assert all(isinstance(result[1], float) for result in results), \
    "Test Failed: k_fold_cross_validation does not return float MSE values"
print("Test Passed: k_fold_cross_validation returns float MSE values")
```

## Generalization error discussion


#### **Concept of Generalization Error**
Generalization error represents the difference between the performance of a model on the training data and unseen test data. It reflects the model's ability to predict correctly for data it has not encountered before. A low generalization error indicates a well-trained model that avoids both underfitting and overfitting.

#### **Implications of \( k = 1 \) in kNN** (Best K chosen by absolute lowest MSE)

   - A small \( k \) (e.g., \( k = 1 \)) makes the kNN model more sensitive to local patterns in the data.
   - While this allows the model to capture finer-grained structures, it also makes it more prone to overfitting, as it may adapt too closely to the noise in the training data.

### Problem 5: Learning Curves
Learning curves were constructed to analyze how the size of the training dataset affected model performance. The training set size was incrementally increased, and the MSE for both the training data and a fixed validation set was calculated at each step. This approach highlighted trends in training and validation errors, providing insights into the model's learning capacity and the potential impact of additional training data. Visualizing these trends offered a clear depiction of the trade-off between underfitting and overfitting.

## compute_learning_curves documentation

Description: Calculates learning curves by incrementally increasing the training set size and computing training and validation errors at each step.  

Parameters:  
data (List[List[float]]): Dataset with features and target values.  
k (int): Number of neighbors for kNN.  
folds (int): Number of folds for validation.  

Returns:  
A tuple containing:  
train_sizes (List[int]): List of training set sizes.  
train_errors (List[float]): Training errors (MSE) for each size.  
val_errors (List[float]): Validation errors (MSE) for each size.  

```python

def compute_learning_curves(data: List[List[float]], k: int, folds: int = 10) -> Tuple[List[int], List[float], List[float]]:
    train_sizes = []
    train_errors = []
    val_errors = []

    # Create folds for cross-validation
    data_folds = create_folds(data, folds)
    
    # Use one fold for validation
    train_data, test_data = create_train_test(data_folds, index=0)
    test_targets = [sample[-1] for sample in test_data]

    # Incrementally increase training set size
    for size in range(50, len(train_data), 50):
        train_subset = train_data[:size]
        train_targets = [sample[-1] for sample in train_subset]

        # Predictions on training data
        train_predictions = [knn_regression(train_subset, sample, k) for sample in train_subset]
        train_error = mean_squared_error_manual(train_targets, train_predictions)

        # Predictions on test data
        val_predictions = [knn_regression(train_subset, sample, k) for sample in test_data]
        val_error = mean_squared_error_manual(test_targets, val_predictions)

        # Append results
        train_sizes.append(size)
        train_errors.append(train_error)
        val_errors.append(val_error)

    return train_sizes, train_errors, val_errors

```

```python
# Testing: compute_learning_curves
print("\nTesting: compute_learning_curves")
train_sizes, train_errors, val_errors = compute_learning_curves(test_data, k=3, folds=2)

# Test 1: Check if train_sizes, train_errors, and val_errors are lists
assert isinstance(train_sizes, list) and isinstance(train_errors, list) and isinstance(val_errors, list), \
    "Test Failed: compute_learning_curves does not return lists"
print("Test Passed: compute_learning_curves returns lists")

# Test 2: Check if train_sizes is increasing
assert train_sizes == sorted(train_sizes), \
    "Test Failed: compute_learning_curves does not return increasing train_sizes"
print("Test Passed: compute_learning_curves returns increasing train_sizes")

# Test 3: Check if errors are floats
assert all(isinstance(error, float) for error in train_errors + val_errors), \
    "Test Failed: compute_learning_curves does not return float errors"
print("Test Passed: compute_learning_curves returns float errors")
```

```python
# Compute learning curves for k=6
train_sizes, train_errors, val_errors = compute_learning_curves(data, k=6)

# Plot the learning curves
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_errors, label="Training Error", marker='o')
plt.plot(train_sizes, val_errors, label="Validation Error", marker='o')
plt.title("Learning Curves: Training vs Validation Error")
plt.xlabel("Training Set Size")
plt.ylabel("Mean Squared Error (MSE)")
plt.legend()
plt.grid()
plt.show()
```
