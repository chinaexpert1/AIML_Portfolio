This notebook demonstrates loading the MovieLens dataset, partitioning it, and visualizing rating and timestamp distributions.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('storage/u.data', sep='\t', names=['user', 'item', 'rating', 'timestamp'])
df.head()
```

```python
# Partition data using stratified sampling by user
train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df['user']
)
print(f"Train size: {len(train_df)}  |  Test size: {len(test_df)}")
```

```python
# Visualize rating distribution
plt.figure(figsize=(8,4))
df['rating'].hist(bins=5)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Rating Distribution')
plt.show()
```

```python
# Visualize timestamp distribution
plt.figure(figsize=(8,4))
df['timestamp'].hist(bins=50)
plt.xlabel('Timestamp')
plt.ylabel('Count')
plt.title('Timestamp Distribution')
plt.show()
```
