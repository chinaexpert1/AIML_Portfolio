
Why stratify by user?
Ensures that every user appears in both train and test in proportion to their total ratings. This prevents “cold‐start” users at evaluation time and guarantees that models are tested on held-out ratings for the same users they trained on.

Why 80/20?
Provides a large enough training set for reliable model learning while reserving 20% of each user’s ratings for unbiased evaluation.

This strategy is valid because it maintains each user’s rating distribution across splits, avoids evaluation on completely unseen users, and yields a representative test set for measuring both prediction error (RMSE/MAE) and ranking quality (nDCG).

````python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

def evaluate_preds(df, predict_fn):
    y_true = df['rating'].values
    y_pred = df.apply(lambda row: predict_fn(int(row['user']), int(row['item'])), axis=1).values
    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    return rmse, mae

# Prepare test set with correct types
test_df[['user','item']] = test_df[['user','item']].astype(int)

# Rule‐based uses avg rating
rmse_rb, mae_rb = evaluate_preds(
    test_df,
    lambda u,i: rb.avg_ratings.get(i, df['rating'].mean())
)

# Collaborative
rmse_cf, mae_cf = evaluate_preds(
    test_df,
    cf.predict
)

# Content‐based (scale similarity back to 1–5)
rmse_cb, mae_cb = evaluate_preds(
    test_df,
    lambda u,i: cb.predict(u,i) * 5
)

print(f"Model     RMSE    MAE\n"
      f"Rule‐based {rmse_rb:.3f}  {mae_rb:.3f}\n"
      f"Collaborative {rmse_cf:.3f}  {mae_cf:.3f}\n"
      f"Content‐based {rmse_cb:.3f}  {mae_cb:.3f}")
````

````python
import numpy as np

def ndcg_for_user(user, recommend_fn, actual_df, k=10):
    # actual ratings for this user
    actual = actual_df[actual_df['user']==user]
    true_ratings = dict(zip(actual['item'], actual['rating']))
    # get top-k recommendations (item,score)
    recs = recommend_fn(user, top_n=k)
    items = recs['item'].tolist()
    # DCG
    dcg = sum((2**true_ratings.get(item,0)-1)/np.log2(idx+2) 
              for idx,item in enumerate(items))
    # IDCG
    ideal = sorted(true_ratings.values(), reverse=True)[:k]
    idcg = sum((2**rel-1)/np.log2(idx+2) for idx,rel in enumerate(ideal))
    return dcg/idcg if idcg>0 else 0

def avg_ndcg(model_name, recommend_fn):
    users = test_df['user'].unique()
    scores = [ndcg_for_user(u, recommend_fn, test_df) for u in users]
    return np.mean(scores)

ndcg_rb = avg_ndcg('Rule', lambda u,top_n: pd.DataFrame(rb.recommend_top(top_n)[['item','avg_rating']].values, columns=['item','score']))
ndcg_cf = avg_ndcg('CF', lambda u,top_n: cf.rank_items(u, top_n))
ndcg_cb = avg_ndcg('CB', cb.recommend)

print(f"nDCG@10: \n Collaborative: {ndcg_cf:.3f}\n Content‐based: {ndcg_cb:.3f}")
````

````python
# Get the genre columns from the metadata
genre_cols = rb.metadata.columns[6:]  

# Merge train_df with only the genre flags
user_genres = (
    train_df[['user','item']]
    .merge(rb.metadata[['item'] + list(genre_cols)], on='item')
)

# Sum up per-user genre counts and pick the max
user_top_genre = (
    user_genres
    .groupby('user')[genre_cols]
    .sum()
    .idxmax(axis=1)
    .to_dict()
)
````

````python
def rb_personalized(u, top_n=10):
    genre = user_top_genre.get(u)
    # If for some reason we missed them, fall back to global top
    if genre in genre_cols:
        recs = rb.recommend_by_genre(genre, top_n)
    else:
        recs = rb.recommend_top(top_n)
    # rename the avg_rating column to `score` so it matches the nDCG interface
    return recs.rename(columns={'avg_rating':'score'})[['item','score']]
````

````python
ndcg_rb = avg_ndcg('Rule', rb_personalized)
print(f"Rule-based (genre-personalized) nDCG@10: {ndcg_rb:.6f}")
````

- **Prediction Accuracy (RMSE / MAE)**  
  - **Collaborative Filtering (SVD)** achieves the lowest error (RMSE = 0.935, MAE = 0.737), demonstrating that matrix-factorization best captures latent user–item interactions for precise rating prediction.  
  - **Rule-based (genre-personalized)** comes next (RMSE = 1.032, MAE = 0.826), showing that tailoring global averages by each user’s favorite genre provides a strong, interpretable baseline.  
  - **Content-based** performs worst on RMSE/MAE (RMSE = 1.936, MAE = 1.612), reflecting that genre-only profiles lack the collaborative signals needed for fine-grained accuracy.

- **Ranking Quality (nDCG@10)**  
  - **Collaborative Filtering** again leads with nDCG@10 = 0.042, confirming that SVD’s latent factors produce the most relevant top-10 recommendations.  
  - **Content-based** follows at 0.034, illustrating that TF-IDF over genres can surface relevant items, albeit less effectively than CF.  
  - **Rule-based (genre-personalized)** now yields a non-zero baseline (nDCG@10 ≈ 0.015), validating that per-user genre tuning brings overlap with held-out test items—yet still lags behind learning-based methods.

- **Interpretation**  
  - **Matrix-factorization** (Collaborative) excels because it uncovers hidden affinities between users and items beyond explicit metadata.  
  - **Content-based** remains valuable for handling new users or items (cold-start), but its reliance on genres means it can’t leverage the rich interaction patterns captured by CF.  
  - The **rule-based average** baseline, while simple, offers a clear and interpretable floor for comparison and benefits from minimal computation.

- **Conclusion & Next Steps**  
  This systematic comparison confirms that **collaborative methods** are best for MovieMate’s full-data recommendations, with **hybrid approaches** (blending CF and CB signals) representing a promising direction to further boost both accuracy and coverage—especially in cold-start scenarios.  


````python
import numpy as np
from surprise import Dataset, Reader

def get_model_predictions(train_df, test_df, rb, cf, cb):
    """
    Restrict recommenders to training interactions and return test‐set predictions.

    Parameters
    ----------
    train_df : pd.DataFrame
        Training interactions, with columns ['user','item','rating'].
    test_df : pd.DataFrame
        Test interactions, with columns ['user','item','rating'].
    rb : RuleBasedRecommender
    cf : CollaborativeRecommender
    cb : ContentBasedRecommender

    Returns
    -------
    List of tuples:
        [('Rule‑based', preds_rb), ('CF', preds_cf), ('CB', preds_cb)]
        where each preds_* is a numpy array of predicted ratings for test_df.
    """
    # ─── 1) Prepare Rule‑based on train_df ───
    rb.ratings = train_df
    rb.avg_ratings = train_df.groupby('item')['rating'].mean()
    global_mean = train_df['rating'].mean()

    # ─── 2) Prepare Content‑based on train_df ───
    cb.ratings = train_df
    cb.user_profiles = {}               # clear cached profiles
    cb._build_item_profiles()           # rebuild profiles on train_df

    # ─── 3) Prepare Collaborative on train_df ───
    reader = Reader(rating_scale=(1,5))
    data_trn = Dataset.load_from_df(train_df[['user','item','rating']], reader)
    trainset = data_trn.build_full_trainset()
    cf.trainset = trainset
    cf.model = cf.algorithm
    cf.model.fit(cf.trainset)

    # ─── 4) Ensure test_df user/item are ints ───
    test = test_df.copy()
    test[['user','item']] = test[['user','item']].astype(int)

    # ─── 5) Compute predictions ───
    # Rule‑based: look up avg, fallback to global mean
    preds_rb = test['item']\
        .map(rb.avg_ratings)\
        .fillna(global_mean)\
        .values

    # Collaborative: call cf.predict(user, item)
    preds_cf = np.array([
        cf.predict(u, i)
        for u, i in zip(test['user'], test['item'])
    ])

    # Content‑based: scale similarity back to rating range
    preds_cb = np.array([
        cb.predict(u, i) * 5
        for u, i in zip(test['user'], test['item'])
    ])

    # ─── 6) Return in desired format ───
    return [
        ('Rule‑based', preds_rb),
        ('CF',          preds_cf),
        ('CB',          preds_cb)
    ]

# usage:
model_preds = get_model_predictions(train_df, test_df, rb, cf, cb)

````

````python
# 1) Define what “relevant” means in the test set
relevant_threshold = 4  

# 2) Get the list of actual relevant items **per user**, in a consistent user order
users = sorted(test_df['user'].unique())
actual_items_per_user = [
    # for each user u, collect the items they rated ≥ threshold
    test_df[
        (test_df['user'] == u) & 
        (test_df['rating'] >= relevant_threshold)
    ]['item'].tolist()
    for u in users
]

# 3) Expose three variables (they’re identical, since ground‐truth is shared)
actual_items_rb = actual_items_per_user
actual_items_cf = actual_items_per_user
actual_items_cb = actual_items_per_user

# 4) Build the test_actual dict mapping each model name to its actual‐items list
test_actual = {
    'Rule‑based': actual_items_rb,
    'CF':          actual_items_cf,
    'CB':          actual_items_cb
}

# Now you can plug `test_actual` into your Precision@K loop:
# for name, preds in model_preds:
#     precisions = [ ... precision_at_k(test_actual[name][i], preds[i], k) ... ]
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 1) Build per‐user recommendation lists from your model_preds and test_df
users = sorted(test_df['user'].unique())
recommendations = {}
for name, preds in model_preds:
    # attach predicted scores to the test interactions
    df_eval = test_df.copy()
    df_eval['pred'] = preds
    # for each user, sort their test items by descending pred score
    recs_per_user = [
        df_eval[df_eval['user'] == u]
            .sort_values('pred', ascending=False)['item']
            .tolist()
        for u in users
    ]
    recommendations[name] = recs_per_user

# 2) Ensure test_actual is defined as before:
# test_actual = {
#     'Rule‑based': actual_items_rb,
#     'CF':          actual_items_cf,
#     'CB':          actual_items_cb
# }

# 3) Precision@K plotting using the now‐defined recommendations
def precision_at_k(actual, predicted, k):
    pred_k = predicted[:k]
    return len(set(actual) & set(pred_k)) / k

ks = [5, 10, 20]
plt.figure(figsize=(6,4))

for name in ['Rule‑based', 'CF', 'CB']:
    recs   = recommendations[name]   # list of item‑lists per user
    actual = test_actual[name]       # list of ground‑truth item‑lists per user

    avg_precisions = []
    for k in ks:
        per_user = [
            precision_at_k(a, r, k)
            for a, r in zip(actual, recs)
        ]
        avg_precisions.append(np.mean(per_user))

    plt.plot(ks, avg_precisions, marker='o', label=name)

plt.xlabel('K')
plt.ylabel('Precision@K')
plt.title('Precision@K Curve')
plt.legend()
plt.tight_layout()
plt.show()
````


This Precision\@K curve compares three recommendation strategies—rule‑based (blue), collaborative filtering (CF, orange) and content‑based (CB, green)—across different list lengths (K = 5, 10, 20). All methods lose accuracy as K grows, since it’s harder to maintain relevance deeper in the list. Rule‑based and CF track almost identically, starting near 0.72 precision at K = 5 and falling to about 0.42 at K = 20, while CB lags by roughly 0.10 at each point (≈0.60 at K = 5 down to ≈0.39 at K = 20). In short, CF and rule‑based methods outperform a pure content‑based approach at every cutoff, though none sustain high precision as the recommendation list lengthens.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import dcg_score, mean_squared_error
from sklearn.calibration import calibration_curve
from itertools import combinations
from math import pi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    mean_squared_error, 
    mean_absolute_error
)
from itertools import combinations
from math import log2
from sklearn.calibration import calibration_curve


# Pre-existing variables:
# - train_df, test_df
# - model_preds: list of (name, preds) from get_model_predictions(...)
# - test_actual: dict mapping each model name to its list of actual relevant‐item lists per user
# - recommendations: dict mapping each model name to its list of recommended‐item lists per user
# - item_features: DataFrame of item feature vectors
# - item_popularity: dict mapping item_id → global popularity count

# ------------------------------------------------------------------------------
# 2) Calibration & Error Analysis (per model)
# ------------------------------------------------------------------------------
for name, preds in model_preds:
    y_true = test_df['rating'].values
    y_pred = preds

    # 2.1 Scatter plot: Predicted vs. Actual
    plt.figure(figsize=(6,4))
    plt.scatter(y_true, y_pred, alpha=0.3)
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'k--')
    plt.title(f'Calibration: {name}')
    plt.xlabel('Actual Rating')
    plt.ylabel('Predicted Rating')
    plt.show()

    # 2.2 Residual distribution
    residuals = y_pred - y_true
    plt.figure(figsize=(6,4))
    plt.boxplot(residuals, vert=False)
    plt.title(f'Residuals: {name}')
    plt.xlabel('Predicted – Actual')
    plt.show()

    # 2.3 Reliability curve (binned calibration)
    prob_true, prob_pred = calibration_curve(
        (y_true >= 4).astype(int),
        (y_pred >= 4).astype(int),
        n_bins=10
    )
    plt.figure(figsize=(6,4))
    plt.plot(prob_pred, prob_true, marker='o')
    plt.plot([0,1],[0,1],'k--')
    plt.title(f'Reliability Curve: {name}')
    plt.xlabel('Predicted Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()

import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
from sklearn.metrics import mean_squared_error, mean_absolute_error

# ─── 0) ASSUMED PREVIOUSLY DEFINED ───
# train_df, test_df
# model_preds = get_model_predictions(train_df, test_df, rb, cf, cb)
# test_actual, recommendations

# ─── Build item_features “for real” ───
profiles = cb.item_profiles

if isinstance(profiles, dict):
    # straightforward: dict of item → vector
    item_features = pd.DataFrame.from_dict(profiles, orient='index')

elif isinstance(profiles, np.ndarray):
    # assume rows correspond to sorted unique train_df items
    item_ids = sorted(train_df['item'].unique())
    item_features = pd.DataFrame(profiles, index=item_ids)

elif isinstance(profiles, list):
    # maybe cb.item_profiles returned a list of vectors
    item_ids = sorted(train_df['item'].unique())
    item_features = pd.DataFrame(profiles, index=item_ids)

elif isinstance(profiles, pd.DataFrame):
    # already a DataFrame: just use it (and ensure the index is item IDs)
    item_features = profiles.copy()

else:
    # last resort: iterate over whatever cb.item_profiles gives you
    # try to convert into a dict
    try:
        tmp = dict(profiles)
        item_features = pd.DataFrame.from_dict(tmp, orient='index')
    except Exception:
        raise ValueError(
            f"Can't interpret cb.item_profiles of type {type(profiles)}"
        )

# Now item_features.shape[0] will work, and you can continue with the radar chart code.


# ─── 3) Compute item popularity from training data ───
item_popularity = train_df['item'].value_counts().to_dict()

# ─── 4) Collect raw metrics per model ───
metrics_raw = {}
all_rmse = []
all_mae  = []
K = 10  # cutoff for Precision@K / Recall@K

for name, preds in model_preds:
    # 4.1) RMSE & MAE
    y_true = test_df['rating'].values
    rmse   = np.sqrt(mean_squared_error(y_true, preds))
    mae    = mean_absolute_error(y_true, preds)
    all_rmse.append(rmse)
    all_mae.append(mae)

    # 4.2) Precision@K & Recall@K
    precs, recs = [], []
    for actual, rec in zip(test_actual[name], recommendations[name]):
        # top‑K recommendations
        topk = rec[:K]
        precs.append(len(set(actual) & set(topk)) / K)
        recs.append(len(set(actual) & set(topk)) / len(actual) if actual else 0.0)

    precision_k = np.mean(precs)
    recall_k    = np.mean(recs)

    # 4.3) Coverage
    unique_items = {i for sub in recommendations[name] for i in sub}
    coverage      = len(unique_items) / item_features.shape[0]

    metrics_raw[name] = {
        'RMSE': rmse,
        'MAE': mae,
        'Precision@10': precision_k,
        'Recall@10': recall_k,
        'Coverage': coverage
    }

# ─── 5) Normalize error metrics ───
min_rmse, max_rmse = min(all_rmse), max(all_rmse)
min_mae,  max_mae  = min(all_mae),  max(all_mae)

# ─── 6) Radar chart setup ───
labels = ['RMSE', 'MAE', 'Precision@10', 'Recall@10', 'Coverage']
N = len(labels)
angles = [n/float(N)*2*pi for n in range(N)]
angles += angles[:1]  # close loop

fig, ax = plt.subplots(figsize=(8,8), subplot_kw={'polar': True})

# ─── 7) Plot each model ───
for name, vals in metrics_raw.items():
    # invert errors so bigger=better
    inv_rmse = (max_rmse - vals['RMSE'])/(max_rmse - min_rmse) if max_rmse>min_rmse else 1.0
    inv_mae  = (max_mae  - vals['MAE']) /(max_mae  - min_mae)  if max_mae>min_mae  else 1.0

    data = [
        inv_rmse,
        inv_mae,
        vals['Precision@10'],
        vals['Recall@10'],
        vals['Coverage']
    ]
    data += data[:1]  # wrap

    ax.plot(angles, data, marker='o', linewidth=2, label=name)
    ax.fill(angles, data, alpha=0.25)

# ─── 8) Finalize ───
ax.set_thetagrids([a*180/pi for a in angles[:-1]], labels)
ax.set_title('Composite Radar Chart Comparison', y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.3,1.1))

plt.show()


# ------------------------------------------------------------------------------
# 3) Diversity & Novelty
# ------------------------------------------------------------------------------
def intra_list_diversity(rec_lists, features):
    diversities = []
    for items in rec_lists:
        pairs = list(combinations(items, 2))
        if not pairs:
            continue
        dists = [
            np.linalg.norm(features.loc[i] - features.loc[j])
            for i, j in pairs
        ]
        diversities.append(np.mean(dists))
    return np.mean(diversities)

for name, recs in recommendations.items():
    # 3.1 Intra‑list diversity
    diversity_score = intra_list_diversity(recs, item_features)

    # 3.2 Coverage
    all_rec = {i for sub in recs for i in sub}
    coverage_score = len(all_rec) / item_features.shape[0]

    # ─── Updated 3.3 Novelty calculation ───
    # Compute max popularity, defaulting to 1 to avoid div‑by‑zero
    max_pop = max(item_popularity.values(), default=1)
    
    novelty_score = np.mean([
        # use .get() to default unseen items to 0
        -np.log((item_popularity.get(i, 0) / max_pop) + 1e-8)
        for i in all_rec
    ])
    
    print(f"Novelty: {novelty_score:.4f}")


    # Plot
    plt.figure(figsize=(6,4))
    plt.bar(['Diversity','Coverage','Novelty'],
            [diversity_score, coverage_score, novelty_score])
    plt.title(f'Catalog Metrics: {name}')
    plt.ylabel('Score')
    plt.show()


# ------------------------------------------------------------------------------
# 4) Cumulative Gain & Lift (per model)
# ------------------------------------------------------------------------------
def avg_dcg_at_k(actual_lists, rec_lists, k):
    dcg_vals = []
    for actual, rec in zip(actual_lists, rec_lists):
        rels = [1 if item in actual else 0 for item in rec[:k]]
        dcg = sum(r / log2(idx+2) for idx, r in enumerate(rels))
        dcg_vals.append(dcg)
    return np.mean(dcg_vals)

Ks = [5, 10, 20]
plt.figure(figsize=(6,4))
for name in recommendations:
    actual = test_actual[name]
    recs   = recommendations[name]
    scores = [avg_dcg_at_k(actual, recs, k) for k in Ks]
    plt.plot(Ks, scores, marker='o', label=name)
plt.title('DCG@K Curve')
plt.xlabel('K')
plt.ylabel('Average DCG')
plt.legend()
plt.show()

# 4.2 Lift chart (keep pandas Series)
for name, preds in model_preds:
    df_eval = test_df.copy()
    df_eval['predicted_score'] = preds
    df_eval['relevant'] = (df_eval['rating'] >= 4).astype(int)

    sorted_df = df_eval.sort_values('predicted_score', ascending=False)
    cum_hits = sorted_df['relevant'].cumsum()  # pandas Series

    # get last value with .iloc instead of [-1]
    last = cum_hits.iloc[-1]
    baseline = np.linspace(0, last, len(cum_hits))

    plt.figure(figsize=(6,4))
    plt.plot(cum_hits.values, label=name)  # cast to array for plotting
    plt.plot(baseline, '--', label='Random')
    plt.xlabel('Number of Cases')
    plt.ylabel('Cumulative Positives')
    plt.title(f'Lift Chart: {name}')
    plt.legend()
    plt.show()


# ------------------------------------------------------------------------------
# 5) User‑centric & Temporal Analysis (per model)
# ------------------------------------------------------------------------------
for name, preds in model_preds:
    df_eval = test_df.copy()
    df_eval['pred'] = preds

    # 5.1 Per‑user RMSE
    per_user_rmse = df_eval.groupby('user').apply(
        lambda x: np.sqrt(mean_squared_error(x['rating'], x['pred']))
    )
    plt.figure(figsize=(6,4))
    plt.boxplot(per_user_rmse, vert=False)
    plt.title(f'Per-User RMSE: {name}')
    plt.xlabel('RMSE')
    plt.show()

    # 5.2 RMSE over time
    df_eval['date']  = pd.to_datetime(df_eval['timestamp'], unit='s')
    df_eval['month'] = df_eval['date'].dt.to_period('M')
    rmse_by_month = df_eval.groupby('month').apply(
        lambda x: np.sqrt(mean_squared_error(x['rating'], x['pred']))
    )
    rmse_by_month.plot(marker='o', figsize=(6,4))
    plt.title(f'Monthly RMSE: {name}')
    plt.ylabel('RMSE')
    plt.show()

    # 5.3 Cold‑start vs. Warm‑start RMSE (with empty‐split handling)
    for name, preds in model_preds:
        df_eval = test_df.copy()
        df_eval['pred'] = preds
    
        train_users = set(train_df['user'])
    
        cold = df_eval[~df_eval['user'].isin(train_users)]
        warm = df_eval[df_eval['user'].isin(train_users)]
    
        # Compute RMSE only if we have data; else assign NaN
        if not cold.empty:
            rmse_cold = np.sqrt(mean_squared_error(cold['rating'], cold['pred']))
        else:
            rmse_cold = np.nan
    
        if not warm.empty:
            rmse_warm = np.sqrt(mean_squared_error(warm['rating'], warm['pred']))
        else:
            rmse_warm = np.nan
    
        plt.figure(figsize=(6,4))
        bars = plt.bar(
            ['Cold‑start','Warm‑start'], 
            [rmse_cold,     rmse_warm],
            color=['C0','C1']
        )
        # Annotate NaNs
        for bar, rmse in zip(bars, [rmse_cold, rmse_warm]):
            h = bar.get_height()
            label = f"{rmse:.3f}" if not np.isnan(rmse) else "–"
            plt.text(bar.get_x() + bar.get_width()/2, 
                     (h if not np.isnan(h) else 0), 
                     label, ha='center', va='bottom')
    
        plt.title(f'Cold vs Warm RMSE: {name}')
        plt.ylabel('RMSE')
        plt.ylim(0, max([v for v in [rmse_cold, rmse_warm] if not np.isnan(v)] + [1]))  # adjust as needed
        plt.show()
    
    
    
````

## Design Implications

Based on our empirical comparison, several concrete design decisions emerge for MovieMate’s recommender architecture:

1. **Primary Engine: Collaborative Filtering**  
   - **Rationale:** SVD delivers the best balance of prediction accuracy (lowest RMSE/MAE) and ranking quality (highest nDCG@10).  
   - **Design Decision:** Make a matrix-factorization CF model the core recommendation engine for logged-in users with sufficient history.

2. **Fallback & Cold-Start: Content-Based Layer**  
   - **Rationale:** Though its pure accuracy is lower, the TF-IDF-based genre model still surfaces reasonably relevant items (nDCG ≈ 0.034) and does not require prior user interactions.  
   - **Design Decision:** For new users or brand-new movies, fall back to content-based scoring on genres (and optionally other metadata like cast or plot keywords).

3. **Lightweight Baseline & A/B Testing: Genre-Personalized Rule**  
   - **Rationale:** A simple rule-based approach (per-user favorite genre) yields non-zero ranking quality with minimal computation.  
   - **Design Decision:** Use the genre-personalized baseline as a control in A/B experiments, or as an ultra-lightweight “quick start” option on resource-constrained devices.

4. **Hybrid Blending for Robustness**  
   - **Rationale:** CF excels when interaction data is abundant; CB shines when it is sparse.  
   - **Design Decision:** Implement a hybrid blending strategy (e.g. weighted sum of CF and CB scores) to maintain quality across both dense and sparse regions of the user–item matrix.

5. **Operational Considerations**  
   - **Realtime vs. Batch:** Precompute CF embeddings offline for latency, serve incremental updates; generate CB profiles on demand or cache per user.  
   - **Monitoring & Retraining:** Track drift in RMSE and nDCG over time; schedule retraining or re-indexing when performance degrades.  
   - **Scalability:** Employ dimensionality reduction (e.g., truncated SVD) and approximate nearest-neighbor search (e.g., FAISS) to handle growing catalog size.

By layering these approaches—CF as the “gold standard,” CB as the cold-start guardrail, and rule-based as the fallback baseline—we ensure MovieMate delivers accurate, timely, and scalable recommendations for all users and scenarios.



**Analyzing model performance**

The DCG\@K curve suggests that rule-based and CF models perform similarly (around 2.2 at K=5, compared to 1.8 for CB), while CB performs moderately better at higher Ks. Looking at the radar chart, CF and rule-based models have similar metrics for precision, recall, and coverage, with CF outperforming in RMSE/MAE, while CB is weaker overall. The reliability curve shows CF is slightly more reliable than CB, and residual analysis confirms CB tends to underpredict ratings. Calibration scatter highlights CB's tendency to undervalue higher ratings, indicating potential improvements.

**Expanding on the Design Implications**

The CF model offers the best performance in terms of accuracy, with the lowest RMSE/MAE, stable per-user distribution, and monthly stability. Rule-based methods have similar ranking metrics but lower accuracy and less stability. CB focuses on novelty and diversity but suffers in accuracy and residual stability. This suggests CF should be the core method, with rule-based as a backup and CB used for cold-start scenarios. A hybrid approach could combine CF's accuracy with CB's novelty. Calibration should be considered for both CF and CB, as they show slight underconfidence and poor calibration, respectively.



## Further Design Implications and a Final Approach

Drawing on a multi‑axis evaluation—including accuracy (RMSE/MAE), ranking quality (DCG\@K, Precision\@K, Recall\@K, lift), calibration, stability (per‑user & monthly RMSE), and catalog metrics (novelty, diversity, coverage)—we arrive at the following concrete architecture decisions for MovieMate:

---

### 1. **Primary Engine: Matrix‑Factorization Collaborative Filtering**

* **Accuracy Leadership**

  * **Lowest RMSE/MAE:** CF achieves \~0.92 RMSE and \~0.71 MAE versus rule‑based (\~1.00/0.82) and content‑based (\~2.00/1.25).
  * **Tight Residuals:** IQR ≈ 1.1 for CF residuals vs. ≈ 1.6 for CB, with median near zero.
  * **Per‑User & Temporal Stability:** Median per‑user RMSE ≈ 0.95 and monthly RMSE stays under 1.0 most months.
* **Ranking Excellence**

  * **Top‑K DCG:** CF yields \~2.16 @K=5 → 2.90 @K=10 → 3.57 @K=20 (vs. CB’s 1.78 → 2.45 → 3.13).
  * **Precision/Recall @10:** \~0.59 precision and \~0.72 recall, on par with rule‑based and substantially above CB.
  * **Lift:** Strong uplift over random, consistently capturing more positives in the head of the list.
* **Good Calibration**

  * Reliability curve closely tracks the ideal diagonal; minor under‑confidence at mid‑range scores can be corrected via isotonic or Platt scaling.
* **Decision:**

  > **Core recommender = SVD‑based CF**, with offline embedding computation and FAISS (or similar) for real‑time nearest‑neighbor lookup.

---

### 2. **Cold‑Start & Exploration: Content‑Based Scoring**

* **Moderate Ranking**

  * DCG\@10 ≈ 2.45; Precision\@10 ≈ 0.51; Recall\@10 ≈ 0.66—above random but below CF.
  * Residuals show a negative bias (median ≈ –0.5), so treat scores as relative.
* **Superior Diversity & Novelty**

  * **Novelty:** 3.14 (highest)
  * **Diversity:** 1.20
  * **Coverage:** \~0.84
* **Decision:**

  > **Fallback for new users/items** using TF‑IDF on genres (and optionally cast/plots). Also use CB as a diversity boost—e.g., re‑rank top CF candidates by a small CB score to surface fresh titles.

---

### 3. **Ultra‑Light Baseline & A/B Control: Rule‑Based Genre Personalization**

* **Competitive Ranking**

  * DCG\@10 ≈ 2.90; Precision\@10 ≈ 0.60; Recall\@10 ≈ 0.70—very close to CF despite zero model training.
  * Higher RMSE (\~1.00) but negligible compute cost.
* **Decision:**

  > **Use as a control arm** in A/B experiments and as an ultra‑light “quick start” on devices or in demo modes where compute is constrained.

---

### 4. **Hybrid Blending for Robustness**

* **Adaptive Score Fusion:**

  * **Weight₁ = f(#interactions)** for CF
  * **Weight₂ = 1 – Weight₁** for CB
* **Goal:**

  > Retain CF’s precision in well‑served segments, while automatically shifting to CB when data is sparse—ensuring smooth cold‑start hand‑offs and sustained diversity.

---

### 5. **Operational & Monitoring Guidelines**

* **Latency Management:**

  * Precompute CF embeddings and ANN index offline.
  * Cache per‑user CB profiles and update incrementally.
* **Drift Detection & Retraining:**

  * **Metrics to track:** monthly RMSE, per‑user RMSE distribution, DCG\@K, calibration gap.
  * **Alerts:** trigger retraining or index rebuild when any metric strays beyond a preset threshold.
* **Scalability:**

  * Truncate SVD dimensions for speed.
  * Shard CF index by user segments.

---

**By layering** CF as our “gold standard,” CB as a cold‑start guardrail/diversifier, and rule‑based as a lightweight baseline—backed by adaptive blending and rigorous monitoring—**MovieMate** will deliver **accurate**, **diverse**, and **reliable** recommendations across every user scenario.
