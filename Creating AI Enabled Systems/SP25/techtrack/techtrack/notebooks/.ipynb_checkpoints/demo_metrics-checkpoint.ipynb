```python
# =============================
# 1. METRICS FUNCTIONS (metrics.py)
# =============================
import numpy as np
from sklearn.preprocessing import label_binarize

def calculate_iou(boxA, boxB):
    """
    Calculate Intersection over Union (IoU) between two bounding boxes.
    Each box is [x, y, w, h] with (x,y) being the top-left corner.
    """
    xA, yA, wA, hA = boxA
    xB, yB, wB, hB = boxB

    x_left = max(xA, xB)
    y_top = max(yA, yB)
    x_right = min(xA + wA, xB + wB)
    y_bottom = min(yA + hA, yB + hB)

    inter_width = max(0, x_right - x_left)
    inter_height = max(0, y_bottom - y_top)
    inter_area = inter_width * inter_height

    boxA_area = wA * hA
    boxB_area = wB * hB
    union_area = boxA_area + boxB_area - inter_area

    iou = inter_area / union_area if union_area > 0 else 0
    return iou

def evaluate_detections(boxes, classes, scores, cls_scores, gt_boxes, gt_classes, map_iou_threshold, eval_type="class_scores"):
    """
    Evaluate detections by matching predictions with ground truth boxes.
    Returns:
        y_true: list of true labels (ground truth or -1 for false positives)
        pred_scores: list of score vectors (one per candidate detection)
    """
    y_true = []
    pred_scores = []
    
    # Determine number of classes
    if cls_scores and isinstance(cls_scores[0], np.ndarray) and cls_scores[0].ndim > 1:
        num_classes = cls_scores[0].shape[1]
    else:
        num_classes = max([max(lst) for lst in gt_classes if lst]) + 1

    for i in range(len(boxes)):
        gt_matched = [False] * len(gt_boxes[i])
        def get_eval_score(j):
            if eval_type == "objectness":
                return scores[i][j]
            elif eval_type == "class_scores":
                return float(max(cls_scores[i][j]))
            elif eval_type == "combined":
                return scores[i][j] * float(max(cls_scores[i][j]))
            else:
                return scores[i][j]
        
        detection_indices = list(range(len(boxes[i])))
        sorted_indices = sorted(detection_indices, key=lambda j: get_eval_score(j), reverse=True)
        
        for j in sorted_indices:
            det_box = boxes[i][j]
            det_class = classes[i][j]
            eval_score = get_eval_score(j)
            
            best_iou = 0
            best_gt_idx = -1
            for k, gt_box in enumerate(gt_boxes[i]):
                iou = calculate_iou(det_box, gt_box)
                if iou > best_iou:
                    best_iou = iou
                    best_gt_idx = k
            
            if best_iou >= map_iou_threshold and best_gt_idx != -1 and (gt_classes[i][best_gt_idx] == det_class) and not gt_matched[best_gt_idx]:
                gt_matched[best_gt_idx] = True
                y_true.append(det_class)
                score_vector = [0] * num_classes
                score_vector[det_class] = eval_score
                pred_scores.append(score_vector)
            else:
                y_true.append(-1)
                score_vector = [0] * num_classes
                score_vector[det_class] = eval_score
                pred_scores.append(score_vector)
        
        # Add false negatives for any unmatched ground truth boxes
        for k, matched in enumerate(gt_matched):
            if not matched:
                y_true.append(gt_classes[i][k])
                score_vector = [0] * num_classes
                pred_scores.append(score_vector)
    
    return y_true, pred_scores

def calculate_precision_recall_curve(y_true, pred_scores, num_classes=20):
    """
    Compute precision-recall curves for each class.
    Returns:
        precision, recall, thresholds: dictionaries keyed by class index.
    """
    precision = {}
    recall = {}
    thresholds = {}
    
    y_true = np.array(y_true)
    pred_scores = np.array(pred_scores)
    
    for c in range(num_classes):
        binary_true = (y_true == c).astype(int)
        scores_c = pred_scores[:, c]
        sorted_indices = np.argsort(scores_c)[::-1]
        sorted_true = binary_true[sorted_indices]
        sorted_scores = scores_c[sorted_indices]
        total_positives = np.sum(binary_true)
        
        prec_vals = []
        rec_vals = []
        thresh_list = []
        unique_thresholds = np.unique(sorted_scores)[::-1]
        
        for thresh in unique_thresholds:
            idx = sorted_scores >= thresh
            TP = np.sum(sorted_true[idx])
            FP = np.sum(1 - sorted_true[idx])
            prec = TP / (TP + FP) if (TP + FP) > 0 else 0
            rec = TP / total_positives if total_positives > 0 else 0
            prec_vals.append(prec)
            rec_vals.append(rec)
            thresh_list.append(thresh)
        
        thresholds[c] = np.append(thresh_list, 0)
        precision[c] = prec_vals
        recall[c] = rec_vals
    
    return precision, recall, thresholds

def calculate_map_x_point_interpolated(precision_recall_points, num_classes, num_interpolated_points=11):
    """
    Compute mAP using x-point interpolation (default 11 points).
    """
    mean_average_precisions = []
    
    for i in range(num_classes):
        points = precision_recall_points[i]
        points = sorted(points, key=lambda x: x[1], reverse=True)
        interpolated_precisions = []
        for recall_threshold in [j * 0.1 for j in range(num_interpolated_points)]:
            possible_precisions = [p for r, p in points if r >= recall_threshold]
            if possible_precisions:
                interpolated_precisions.append(max(possible_precisions))
            else:
                interpolated_precisions.append(0)
        mean_average_precision = sum(interpolated_precisions) / len(interpolated_precisions)
        mean_average_precisions.append(mean_average_precision)
    
    overall_map = sum(mean_average_precisions) / num_classes
    return overall_map

# =============================
# 2. DETECTION LOADING FUNCTIONS
# =============================
import os

# Define predictions folder constant
PREDICTIONS_FOLDER = "/detections"

def load_detections(pred_folder: str, num_classes: int):
    """
    Loads detection files from the predictions folder.
    Each file is expected to contain lines formatted as:
        x y w h objectness cls_score_0 cls_score_1 ... cls_score_19
    Returns:
        boxes: List of detected bounding boxes per image (each is a list of floats [x, y, w, h]).
        classes_list: List of predicted class IDs (computed from the class score vector).
        scores_list: List of objectness scores.
        cls_scores_list: List of class score vectors (list of floats).
    """
    detection_files = sorted([f for f in os.listdir(pred_folder) if f.lower().endswith('.txt')])
    
    boxes = []
    classes_list = []
    scores_list = []
    cls_scores_list = []
    
    for file in detection_files:
        file_path = os.path.join(pred_folder, file)
        with open(file_path, "r") as f:
            lines = f.readlines()
        
        image_boxes = []
        image_classes = []
        image_scores = []
        image_cls_scores = []
        
        for line in lines:
            parts = line.strip().split()
            if len(parts) != 25:
                continue  # Skip malformed lines
            bbox = list(map(float, parts[0:4]))
            obj_score = float(parts[4])
            cls_scores = list(map(float, parts[5:]))
            pred_class = int(np.argmax(cls_scores))
            
            image_boxes.append(bbox)
            image_classes.append(pred_class)
            image_scores.append(obj_score)
            image_cls_scores.append(cls_scores)
        
        boxes.append(image_boxes)
        classes_list.append(image_classes)
        scores_list.append(image_scores)
        cls_scores_list.append(image_cls_scores)
    
    return boxes, classes_list, scores_list, cls_scores_list

def load_ground_truth(gt_folder: str):
    """
    Loads ground truth files from the given folder.
    Each file should contain lines formatted as: 'class x y w h'
    where the bounding box coordinates (x, y, w, h) are floats.
    Returns:
        gt_boxes: List of ground truth bounding boxes per image (each is a list of floats [x, y, w, h]).
        gt_classes: List of ground truth class labels per image.
    """
    gt_files = sorted([f for f in os.listdir(gt_folder) if f.lower().endswith('.txt')])
    
    gt_boxes = []
    gt_classes = []
    
    for file in gt_files:
        file_path = os.path.join(gt_folder, file)
        with open(file_path, "r") as f:
            lines = f.readlines()
        
        image_gt_boxes = []
        image_gt_classes = []
        
        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
            cls_id = int(parts[0])
            bbox = list(map(float, parts[1:]))
            image_gt_boxes.append(bbox)
            image_gt_classes.append(cls_id)
        
        gt_boxes.append(image_gt_boxes)
        gt_classes.append(image_gt_classes)
    
    return gt_boxes, gt_classes

# =============================
# 3. RUNNER CODE
# =============================
import matplotlib.pyplot as plt

# Specify folder paths (adjust as needed)
PREDICTIONS_FOLDER = "../detections"          # Predictions folder (relative path)
GROUND_TRUTH_FOLDER = "../storage/logistics"    # Ground truth folder (relative path)

# Load class names to determine number of classes
with open("../storage/yolo_models/logistics.names", "r") as f:
    class_names = [line.strip() for line in f.readlines()]
num_classes = len(class_names)
print(f"Number of classes: {num_classes}")
print("Class names:", class_names)

# Load predictions and ground truth data using the new format
pred_boxes, pred_classes, pred_scores, pred_cls_scores = load_detections(PREDICTIONS_FOLDER, num_classes)
gt_boxes, gt_classes = load_ground_truth(GROUND_TRUTH_FOLDER)

print(f"Loaded detections for {len(pred_boxes)} images.")
print(f"Loaded ground truth for {len(gt_boxes)} images.")

# Set IoU threshold for evaluation
map_iou_threshold = 1e-05

# Evaluate detections:
y_true, pred_score_vectors = evaluate_detections(
    pred_boxes, pred_classes, pred_scores, pred_cls_scores,
    gt_boxes, gt_classes, map_iou_threshold, eval_type="class_scores"
)




```

#### **Task A:**
**Display the precision-recall curve for each of the classes.**

```python
# Compute precision-recall curves per class
precision, recall, thresholds_dict = calculate_precision_recall_curve(
    y_true, pred_score_vectors, num_classes=num_classes
)

# Create precision-recall pairs for each class
precision_recall_points = {c: list(zip(recall[c], precision[c])) for c in range(num_classes)}



# Plot precision-recall curves for each class
for c in range(num_classes):
    plt.figure()
    plt.plot(recall[c], precision[c], marker='o')
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"Precision-Recall Curve for Class {c}: {class_names[c]}")
    plt.grid(True)
    plt.show()
```

#### **Task B:**
**Calculate the Average Precision of each class.**

```python
def calculate_average_precisions(precision_recall_points, num_interpolated_points=11):
    """
    Calculate the Average Precision (AP) for each class using 11-point interpolation.
    
    For each class, this function interpolates the precision at recall thresholds
    0.0, 0.1, ..., 1.0. For each recall threshold, it takes the maximum precision for 
    all (recall, precision) pairs where the recall is greater than or equal to the threshold.
    
    Parameters
    ----------
    precision_recall_points : dict
        A dictionary where each key is a class index and each value is a list of (recall, precision)
        tuples for that class.
    num_interpolated_points : int, optional
        The number of recall thresholds to use (default is 11 for thresholds 0.0, 0.1, , 1.0).
    
    Returns
    -------
    average_precisions : dict
        A dictionary where each key is a class index and the corresponding value is the Average Precision for that class.
    """
    average_precisions = {}
    
    for c, points in precision_recall_points.items():
        interpolated_precisions = []
        for recall_threshold in [j * 0.1 for j in range(num_interpolated_points)]:
            # Gather all precision values where recall >= threshold
            possible_precisions = [p for (r, p) in points if r >= recall_threshold]
            if possible_precisions:
                interpolated_precisions.append(max(possible_precisions))
            else:
                interpolated_precisions.append(0)
        average_precisions[c] = sum(interpolated_precisions) / num_interpolated_points
    
    return average_precisions

# Example usage in your runner code:
# (Assuming you have already computed precision_recall_points from your predictions)
average_precisions = calculate_average_precisions(precision_recall_points)
for c in range(len(average_precisions)):
    print(f"Class {c} ({class_names[c]}): AP = {average_precisions[c]:.4f}")
```
