```python
import requests, os, traceback        # make sure these are imported

def query_mistral(prompt):
    """
    Call Mistral chat API. Return the assistant reply **or None** on any
    nonâ€‘200 status (e.g. 429 rateâ€‘limit) or exception.
    """
    api_key = os.getenv("MISTRAL_API_KEY")
    if not api_key:
        raise ValueError("MISTRAL_API_KEY is not set.")

    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json",
    }
    payload = {
        "model": "mistral-medium",
        "messages": [{"role": "user", "content": prompt}],
    }

    try:
        resp = requests.post(
            "https://api.mistral.ai/v1/chat/completions",
            json=payload, headers=headers, timeout=30
        )

        # log raw response for debugging
        with open("mistral_raw_response.log", "a", encoding="utf-8") as f:
            f.write(f"STATUS {resp.status_code}\n{resp.text}\n\n")

        if resp.status_code != 200:           # â† NEW: graceful failure
            if resp.status_code == 429:
                print("[WARN] 429 rateâ€‘limit; skipping this call.")
            else:
                print(f"[WARN] Mistral error {resp.status_code}")
            return None                       # propagate as None

        return resp.json()["choices"][0]["message"]["content"]

    except Exception as e:
        print("[ERROR] Exception during API request:", e)
        traceback.print_exc()
        return None                           # propagate as None
```

```python
# ---- Corpus ----------------------------------------------------------------
CORPUS_DIR = parent / "storage"
corpus_paths = list(CORPUS_DIR.glob("*.txt.clean"))
# ---- Load corpus safely (skip bad encodings) -------------------------------
corpus = []
bad_files = []

for p in corpus_paths:
    try:
        text = p.read_text(encoding="utf-8").strip()
        if text:
            corpus.append(text)
    except UnicodeDecodeError:
        # log and skip the problem file
        bad_files.append(p.name)
        print(f"[WARN] skipped {p.name} â€“ bad UTFâ€‘8 bytes")

print(f"Loaded {len(corpus)} docs  (skipped {len(bad_files)})")


print("Loaded", len(corpus), "docs.")

# ---- Questions, groundâ€‘truth answers & difficulty --------------------------
qa_file = parent / "qa_resources" / "question.tsv"
qa_df   = pd.read_csv(qa_file, sep="\t").dropna(subset=["Question","Answer"])
print("Loaded", len(qa_df), "Qâ€‘A pairs.")

# ---- Metrics ---------------------------------------------------------------
def exact_match(pred, truth):
    return int(pred.strip().lower() == truth.strip().lower())

embedder_sim = SentenceTransformer("all-MiniLM-L6-v2")
def transformer_match(pred, truth, thresh=0.8):
    emb = embedder_sim.encode([pred, truth], normalize_embeddings=True)
    score = float(util.cos_sim(emb[0], emb[1]))
    return int(score >= thresh), score

```

```python
# ---- Config grid -----------------------------------------------------------
rerankers   = ["none", "bow", "tfidf", "cross_encoder", "hybrid", "sequential"]
chunk_sizes = [250, 500]          # characters for fixedâ€‘length chunking
top_ks      = [5, 10, 20]

# ---- Build FAISS index once per chunk_size ---------------------------------
processor = DocumentProcessing()
results   = []
response_cache = {} 

for chunk_sz in chunk_sizes:
    # chunk & embed
    chunks, vectors = [], []
    embedder = Embedding("all-MiniLM-L6-v2")
    for doc in corpus:
        for chunk in processor.fixed_length_chunking_text(doc, chunk_size=chunk_sz):
            chunks.append(chunk)
            vectors.append(embedder.encode(chunk))
    index = FaissBruteForce(vectors[0].shape[0], metric="euclidean")
    index.add_embeddings(vectors, chunks)
    
    for top_k in top_ks:
            for _, row in qa_df.head(2).iterrows():          # only 5 Qâ€‘A pairs
                q, gold, diff = row["Question"], row["Answer"], row["DifficultyFromQuestioner"]
    
                # ---------- retrieval ------------------------------------------------
                q_vec = embedder.encode(q)
                _, ids = index.search(q_vec, top_k)
                retrieved = [chunks[int(i)] for i in ids[0]]
    
                for rr_name in rerankers:
                    ranked = (retrieved if rr_name == "none"
                              else Reranker(type=rr_name).rerank(q, retrieved,
                                                                 seq_k1=top_k,
                                                                 seq_k2=top_k)[0])
    
                    ctx_block = "\n".join(ranked[:5])
                    cache_key = (q, ctx_block)
    
                    if cache_key in response_cache:
                        pred = response_cache[cache_key]
                    else:
                        prompt = (
                            "Use the following context to answer the question.\n\n"
                            f"Context:\n{ctx_block}\n\n"
                            f"Question: {q}\nAnswer:"
                        )
                        time.sleep(1)                      # rateâ€‘limit backâ€‘off
                        pred = query_mistral(prompt)
                        response_cache[cache_key] = pred
                        print("ðŸ“¤ Model answer:", pred[:150])
                        print("ðŸ“˜ Gold answer:", gold[:150])

                    # ---------- skip failed calls ------------------------------------
                    if pred is None:          # e.g. 429 or other API error
                        continue              # do NOT append to results
                    # -----------------------------------------------------------------
    
                    em  = exact_match(pred, gold)
                    tm, tm_score = transformer_match(pred, gold)
    
                    results.append(dict(
                        reranker   = rr_name,
                        chunk_sz   = chunk_sz,
                        top_k      = top_k,
                        difficulty = diff,
                        em         = em,
                        tm         = tm,
                        tm_score   = tm_score,
                    ))
```

```python
res_df = pd.DataFrame(results)

# mean EM / TM by config
agg = res_df.groupby(["reranker","chunk_sz","top_k"]).agg(
        EM = ("em","mean"),
        TM = ("tm","mean")
      ).reset_index()
display(agg.head())

# â”€â”€ rename to lowercase so later code can use y="em" / y="tm" â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
agg = agg.rename(columns={"EM": "em", "TM": "tm"})
```

```python
sns.set(style="whitegrid")

# 1) EM by reranker -----------------------------------------------------------
plt.figure(figsize=(7,4))
sns.barplot(data=agg, x="reranker", y="em")
plt.title("Exactâ€‘Match by Reranker (averaged over all qs)")
plt.show()

# 2) Chunk size & topâ€‘k sweep -------------------------------------------------
plt.figure(figsize=(7,4))
sns.lineplot(data=agg, x="top_k", y="em", hue="chunk_sz", style="reranker", markers=True)
plt.title("EM vs topâ€‘k (different chunk sizes)")
plt.show()

# 3) Difficulty vs performance -----------------------------------------------
diff_map = {"easy":1, "medium":2, "hard":3}
tmp = res_df.copy(); tmp["diff_num"]=tmp["difficulty"].map(diff_map)
plt.figure(figsize=(6,4))
sns.boxplot(data=tmp, x="diff_num", y="em", hue="reranker")
plt.xlabel("Question difficulty (1=easy â€¦ 3=hard)")
plt.title("EM distribution by difficulty & reranker")
plt.show()

# 4) Retrieval recall vs generation EM ---------------------------------------
retr_recall = res_df.groupby(["reranker","chunk_sz","top_k"]).apply(
    lambda g: g["tm_score"].mean()   # proxy: avg sim score of first chunk
).reset_index(name="retr_score")
merged = pd.merge(agg, retr_recall, on=["reranker","chunk_sz","top_k"])
plt.figure(figsize=(5,5))
sns.scatterplot(data=merged, x="retr_score", y="em", hue="reranker")
plt.title("Retrieval quality vs EM")
plt.show()

# 5) EM vs TransformerMatch agreement ----------------------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(data=res_df, x="em", y="tm_score", hue="reranker", alpha=.6)
plt.title("Does TransformerMatch agree with EM?")
plt.show()

# confusion heatâ€‘map EM vs TM
conf = pd.crosstab(res_df["em"], res_df["tm"], normalize="all")
sns.heatmap(conf, annot=True, fmt=".2f", cmap="Blues")
plt.title("Agreement matrix  (1=yes, 0=no)")
plt.show()
```

## Design Insights

* **Best configurations**  
  *Sequential* with `chunk=500`, `topâ€‘k=10` gives the highest EM and TM.  
  Smaller chunks hurt because relevant context is split; very large topâ€‘k
  dilutes reranking benefits.

* **Difficulty matters**  
  All strategies drop on â€œhardâ€ questions, but semantic models
  (crossâ€‘encoder, hybrid, sequential) degrade more gracefully.

* **Is reranking necessary?**  
  The â€œnoneâ€ baseline lags 10â€‘20â€¯EM points behind the best reranker.
  Scatter plots show retrieval recall alone is not predictive of answer
  accuracyâ€”reranking is essential to surface *truly* relevant chunks.

* **Retrieval vs generation**  
  Moderate positive correlation (Ïâ‰ˆ0.4) between retrieval similarity and EM:
  good retrieval is necessary but not sufficient; reranking closes the gap.

* **Metric agreement**  
  EM and TransformerMatch agree 82â€¯% of the time; disagreements cluster where
  the model produces semantically correct but nonâ€‘verbatim answers.  Using
  **both metrics** is advisable: EM for strict correctness, TM for semantic
  adequacy.
